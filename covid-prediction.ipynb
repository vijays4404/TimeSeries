```python
#importing library
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
#loading the dataset
df=pd.read_csv('../input/covid19-healthy-diet-dataset/Food_Supply_Quantity_kg_Data.csv')
df
```

```python
df.info()
```

```python
#Preprocessing
df=df.drop('Unit (all except Population)',axis=1)
```

```python
#checking for null values
df.isna().sum()
```

```python
#filling the missing 
for column in df.columns:
    if df.dtypes[column]!='object' and df.isna().sum()[column]>0:
        df[column]=df[column].fillna(df[column].mean())
```

```python
#checking for number of 
df['Undernourished'].value_counts()
```

```python
undernourished_numeric=df.loc[df['Undernourished']!='<2.5','Undernourished']
```

```python
undernourished_numeric
```

```python
undernourished_numeric=undernourished_numeric.astype(np.float)
```

```python
undernourished_numeric=undernourished_numeric.fillna(undernourished_numeric.mean())
```

```python
undernourished_numeric=pd.qcut(undernourished_numeric,q=4,labels=[1,2,3,4])
undernourished_numeric
```

```python
undernourished_numeric.index
```

```python
df['Undernourished']=df['Undernourished'].apply(lambda x: 0 if x=='<2.5' else x)
```

```python
df.loc[undernourished_numeric.index,'Undernourished']=undernourished_numeric
```

```python
df
```

```python
df['Undernourished']=df['Undernourished'].apply(lambda x:0 if x=='<2.5' else x)
```

```python
df
```

```python
df=df.drop('Country',axis=1)
df=df.drop(['Confirmed','Recovered','Active'],axis=1)
```

```python
df['Death']=pd.qcut(df['Deaths'],q=2,labels=[0,1])
```

```python
df
```

```python
y=df['Death']
x=df.drop('Death',axis=1)
```

```python
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns,index=x.index)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
x.shape
```

```python
inputs=tf.keras.Input(shape=(27,))
x=tf.keras.layers.Dense(64,activation='relu')(inputs)
x=tf.keras.layers.Dense(64,activation='relu')(x)
outputs=tf.keras.layers.Dense(1,activation='sigmoid')(x)
                        
model=tf.keras.Model(inputs=inputs,outputs=outputs)
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=[
        'accuracy',
        tf.keras.metrics.AUC(name='auc')
    ])
batch_size=64
epochs=14
history=model.fit(
    x_train,
    y_train,
    validation_split=0.2,
    batch_size=batch_size,
    epochs=epochs,
    verbose=0)
```

```python
fig=px.line(
    history.history,
    y=['loss','val_loss'],
    labels={'index':'Epoch','value':'Loss'},
    title='Training and validation loss')
fig.show()
```

```python
model.evaluate(x_test,y_test)
```
