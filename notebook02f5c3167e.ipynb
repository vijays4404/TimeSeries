```python
#importing library
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('../input/world-university-rankings/cwurData.csv')
df
```

```python
df.drop('institution',axis=1,inplace=True)
```

```python
df
```

```python
df.drop('year',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
#droping broad_impact
df.drop('broad_impact',axis=1,inplace=True)
```

```python
label=LabelEncoder()
df['country']=label.fit_transform(df['country'])
df
```

```python
country_mapping={index:label for index,label in enumerate(label.classes_)}
country_mapping
```

```python
y=df['world_rank']
x=df.drop('world_rank',axis=1)
```

```python
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
