```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#loading the dataset
df=pd.read_csv('/kaggle/input/top-women-chess-players/top_women_chess_players_aug_2020.csv')
#loading the dataset
df
```

```python
#Preprocessing the dataset

#dropping the fide id and name


df=df.drop(['Fide id','Name','Gender'],axis=1)
df
```

```python
#Missing values

df.isna().sum()
```

```python
#checking for datatype

df.dtypes
```

```python
numerical_columns=[column for column in df.columns if df[column].dtypes!='object']
```

```python
numerical_columns
```

```python
for column in numerical_columns:
    df[column]=df[column].fillna(df[column].mean())
```

```python
df['Title'].unique()
```

```python
df['Inactive_flag'].unique()
```

```python
df['Inactive_flag']=df['Inactive_flag'].fillna('wa')
```

```python
df
```

```python
title_dummies=pd.get_dummies(df['Title'])
```

```python
df=pd.concat([df,title_dummies['GM']],axis=1)
```

```python
#dropping the title column
df=df.drop('Title',axis=1)
```

```python
df['GM']=df['GM'].astype(int)
```

```python
df
```

```python
encoder=LabelEncoder()
df['Inactive_flag']=encoder.fit_transform(df['Inactive_flag'])
```

```python
#dropping the federation column
df=df.drop('Federation',axis=1)
```

```python
df
```

```python
#splitting the dataset
y=df['GM']
x=df.drop('GM',axis=1)
```

```python
x
```

```python
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

```python
#train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=123)
```

```python
#train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
print('The model accuracy is',model.score(x_test,y_test))
```
