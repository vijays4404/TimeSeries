# Extracting year and month from date column

```python
def get_year(df,column):
    return df[column].apply(lambda x:x[0:4])
    

def get_month(df,column):
    return df[column].apply(lambda x:x[5:7])
    
    
```

```python
df['Start_Time']
```

```python
df['Start_Year']=get_year(df,'Start_Time')
df['Start_Month']=get_month(df,'Start_Time')
```

```python
df['End_Year']=get_year(df,'End_Time')
df['End_Month']=get_month(df,'End_Time')
df['Weather_Year']=get_year(df,'Weather_Timestamp')
df['Weather_Month']=get_month(df,'Weather_Timestamp')
```

# Create onehot_encode function to encode the object columns

```python
def onehot_encode(df,columns,prefixes):
    df=df.copy()
    for column,prefix in zip(columns,prefixes):
        dummies=pd.get_dummies(df[column],prefix=prefix)
        #concating
        df=pd.concat([df,dummies],axis=1)
        #dropping original column
        df=df.drop(column,axis=1)
    return df
```

```python
df.columns
```

```python
date_column=['Start_Year','Start_Month','End_Year','End_Month','Weather_Year','Weather_Month']
for column in date_column:
    df[column]=df[column].astype(np.int)
```

```python
categorical_columns=[column for column in df.columns if df[column].dtypes=='object']
```

```python
df=onehot_encode(df,categorical_columns,prefixes=categorical_columns)
```
