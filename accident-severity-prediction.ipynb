```python
#columns to drop because of high number of missing values
null_columns=['End_Lat','End_Lng','Number','Wind_Chill(F)','Precipitation(in)']
#dropping columns
df=df.drop(null_columns,axis=1)
```

```python
#number of columns with high number of unique values
unneeded_columns=['ID','Description','Street','City','Zipcode','Country']
df=df.drop(unneeded_columns,axis=1)
```

```python
df
```

```python
#creating function to extract year and month
def get_year(df,column):
    return df[column].apply(lambda df:df[0:4])
def get_month(df,column):
    return df[column].apply(lambda df:df[5:7])
```

```python
#extracing month and year from the date column
df['Start_Time_Year']=get_year(df,'Start_Time')
df['Start_Time_Month']=get_month(df,'Start_Time')

df['End_Time_Year']=get_year(df,'End_Time')
df['End_Time_Month']=get_month(df,'End_Time')

df['Weather_Timestamp_Year']=get_year(df,'Weather_Timestamp')
df['Weather_Timestamp_Month']=get_month(df,'Weather_Timestamp')







```

```python
#removing origin date column
unneeded_column=['Start_Time','End_Time',"Weather_Timestamp"]
#dropping
df=df.drop(unneeded_column,axis=1)
```

```python
#creating function for onehot_encode
def onehot_encode(df,columns,prefixes):
    df=df.copy()
    #creating loop through list of columns
    for column,prefix in zip(columns,prefixes):
        #using get_dummies method  and save it to a variable
        dummies=pd.get_dummies(df[column],prefix=prefix)
        #using concat method to add column to the dataset
        df=pd.concat([df,dummies],axis=1)
        #dropping the original column
        df=df.drop(column,axis=1)
    return df
```

```python
{column:len(df[column].unique()) for column in df.columns if df.dtypes[column]=='object'}
```

```python
#calling the function
df=onehot_encode(df,
                columns=['Side','County','State','Timezone',
                        'Airport_Code','Wind_Direction','Weather_Condition'],
                prefixes=['si','co','st','t2','ac','wd','wc'])
```

```python
#creating binary encoding function to convert it in 0 and 1
def get_binary_column(df,column):
    return df[column].apply(lambda x:1 if x=='Day' else 0)
```

```python
df.info()
```

```python
df['Sunrise_Sunset']=get_binary_column(df,'Sunrise_Sunset')
df['Civil_Twilight']=get_binary_column(df,'Civil_Twilight')
df['Nautical_Twilight']=get_binary_column(df,'Nautical_Twilight')
df['Astronomical_Twilight']=get_binary_column(df,'Astronomical_Twilight')
```

```python
df
```

```python
#Spliting 
y=df['Severity']
x=df.drop('Severity',axis=1)
```

```python
y.unique()
```

```python
y=y-2
```

```python
y.unique()
```

```python
y
```

```python
#scaling the dataset
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
x
```

```python
#train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=100)
```

```python
x.shape
```

```python
#creating layer for tensorflow
inputs=tf.keras.Input(shape=(x.shape[1],))
x=tf.keras.layers.Dense(64,activation='relu')(inputs)
x=tf.keras.layers.Dense(64,activation='relu')(x)
outputs=tf.keras.layers.Dense(4,activation='softmax')(x)
```

```python
model=tf.keras.Model(inputs,outputs)
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'])
batch_size=32
epochs=20
history=model.fit(
    x_train,
    y_train,
    validation_split=0.2,
    batch_size=batch_size,
    epochs=epochs,
    callbacks=[
        tf.keras.callbacks.ReduceLROnPlateau(),
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=3,
            restore_best_weights=True)])
```

```python
print('Test Accuracy',model.evaluate(x_test,y_test))
```
