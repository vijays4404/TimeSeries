```python
#importing library
import numpy as np
import pandas as pd
#train_test_split
from sklearn.model_selection import train_test_split
#StandardScaler
from sklearn.preprocessing import StandardScaler
#LogisticRegression
from sklearn.linear_model import LogisticRegression
#neighbors
from sklearn.neighbors import KNeighborsClassifier
#Decision
from sklearn.tree import DecisionTreeClassifier
#Support Vector Machine
from sklearn.svm import LinearSVC,SVC
#neural_network
from sklearn.neural_network import MLPClassifier
#RandomForest
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
import warnings
warnings.filterwarnings(action='ignore')
```

# Loading the Dataset

```python
df=pd.read_csv('/kaggle/input/audit-data/audit_data.csv')
#showing the dataset
df
```

# Preprocessing Inputs

```python
def preprocess_inputs(df):
    df=df.copy()
    #filling the missing value with mean of that column
    df['Money_Value']=df['Money_Value'].fillna(df['Money_Value'].mean())
    
    #converting quantitative data to onehot_column
    dummies=pd.get_dummies(df['LOCATION_ID'],prefix='LOCATION_ID')
    df=pd.concat([df,dummies],axis=1)
    
    df=df.drop('LOCATION_ID',axis=1)
    
    #splitting between target and feature column
    
    y=df['Risk']
    x=df.drop('Risk',axis=1)
    
    #train_test_split
    
    x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
    
    #scaler
    
    scaler=StandardScaler()
    
    scaler.fit(x_train)
    
    x_train=pd.DataFrame(scaler.transform(x_train),columns=x_train.columns)
    x_test=pd.DataFrame(scaler.transform(x_test),columns=x_test.columns)
    
    
    
    
    
    return x_train,x_test,y_train,y_test
```

```python
x_train,x_test,y_train,y_test=preprocess_inputs(df)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Training the Model

```python
models={
    "LogisticRegression":LogisticRegression(),
    
    'K Neighbors Classifier':KNeighborsClassifier(),
    'Decision Tree Classifier':DecisionTreeClassifier(),
    'Support Vector Machine':SVC(),
    'Support Vector Machine Linear Kernel':LinearSVC(),
    'MLP Classifier':MLPClassifier(),
    'Random Forest': RandomForestClassifier(),
    
    'Gradient Boosting':GradientBoostingClassifier(),

}
```

```python
for name,model in models.items():
    model.fit(x_train,y_train)
    print(name)
    print('Root Mean Square Error',np.sqrt(np.mean((model.predict(x_test)-y_test)**2)))
    print('R2 Score',model.score(x_test,y_test))
```
