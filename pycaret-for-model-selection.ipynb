```python
#importing library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pycaret.classification import setup,models,compare_models,predict_model
```

```python
#loading the dataset
df=pd.read_csv('../input/glass/glass.csv')
#looking into the dataset
df
```

```python
#getting information about the dataset
df.info()
```

```python
#checking for corelation
corr=df.corr()
plt.figure(figsize=(16,15))
sns.heatmap(corr,annot=True,vmin=-1.0,vmax=1,cmap='mako')
plt.title('Correlation')
plt.show()
```

```python
df['Type'].value_counts()
```

```python
plt.figure(figsize=(12,12))
plt.pie(df['Type'].value_counts(),labels=df['Type'].value_counts().index,colors=sns.color_palette('mako'))
plt.title('Class Distribution')
plt.show()
```

# Model Comparison and Evaluation

```python
#setting up pycaret
setup(
    data=df,
    target='Type',
    normalize=True,
    train_size=0.7,
    )
```

```python
#comparing different model and finding best one
best_model=compare_models()
```

```python
#selecting best model based on performance
predict_model(best_model)
```
