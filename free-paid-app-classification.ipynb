```python
#importing  library
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression,LogisticRegressionCV
```

```python
#loading the dataset
df=pd.read_csv('../input/apple-store-ranks-2019/ranks.csv')
#showing the dataset
df
```

```python
#getting information about the dataset
df.info()
```

```python
#iunneeded columns
unneeded_column=['appid','name','publisher']
df=df.drop(unneeded_column,axis=1)
```

```python
#checking for null values 
df.isna().sum()
```

```python
 #checking for missing values
print('Total Missing Values',df.isna().sum().sum())
```

```python
df['year']=df['date'].apply(lambda x:np.int(x[0:4]))
df['month']=df['date'].apply(lambda x:np.int(x[5:7]))
df['day']=df['date'].apply(lambda x:np.int(x[-2:]))

```

```python
#dropping the original day column
df=df.drop('date',axis=1)
```

```python
df
```

```python
df['category'].unique()
```

```python
#creating dummies
category_dummies=pd.get_dummies(df['category'])
```

```python
#concating the dummies with original column
df=pd.concat([df,category_dummies],axis=1)
```

```python
#dropping original category columns
df=df.drop('category',axis=1)
```

```python
#dropping the grossing index
df=df.drop(grossing_index,axis=0).reset_index(drop=True)
```

# #Splitting and Scaling the DataSet

```python
#spliting the dataframe to x and y
y=df['feed'].copy()
x=df.drop('feed',axis=1).copy()
```

```python
#scaling the dataset
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
#train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=34)
```

```python
#training the model
#creating instance of LogisticRegression class
base_model=LogisticRegression()
#apply attributes to the instances base_model
base_model.fit(x_train,y_train)
base_acc=base_model.score(x_test,y_test)
```

```python
#PRINTING THE ACCURACY
print('Accuracy:{:4f}'.format(base_acc))
```

```python
#creating new instance
cv_model=LogisticRegressionCV()
cv_model.fit(x_train,y_train)
cv_acc=cv_model.score(x_test,y_test)
print('Accuray:{:.4f}',format(cv_acc))
```
