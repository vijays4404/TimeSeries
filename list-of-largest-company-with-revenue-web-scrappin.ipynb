```python
#importing the necessary library
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# Scrapping the list of Company with largest revenue

```python
#saving the address to a variable
url='https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
```

```python
#using get method from requests library to pull the data
page=requests.get(url)

soup=BeautifulSoup(page.text,'html')
#printing the data

```

```python
table=soup.find_all('table')[0]
```

```python
world_titles=table.find_all('th')
```

```python
world_titles=[title.text.strip() for title in world_titles]
```

```python
world_titles=world_titles[:7]
```

```python
world_titles.append('S.No')
```

```python
df=pd.DataFrame(columns=world_titles)
```

```python
df.drop('Rank',axis=1)
```

```python
list=[]
i=0
for text in table.find_all('td'):
    text=text.text
    list.append(text)

list=[item.replace('\n','') for item in list]
list=[item.strip() for item in list]
```

```python
list=[item for item in list if item!='']
list=[item for item in list if item[0]!='[']
print(list)
```

```python
# Reshape the list
num_columns = len(df.columns)
reshaped_data = [list[i:i+num_columns] for i in range(0, len(list), num_columns)]

# Assign the reshaped list to DataFrame columns
df = pd.DataFrame(reshaped_data, columns=df.columns)
df
df = df.style.set_properties(**{'text-align': 'mid'})
```
