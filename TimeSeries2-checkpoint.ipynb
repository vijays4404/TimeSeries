```python
def print_chart(csv,title,col):
    df_1=pd.read_csv(csv,index_col='Date',parse_dates=True)
    f_1=plt.figure(figsize=(12,6),dpi=100)
    ax_1=f_1.add_axes([0.0,0.0,0.9,0.9])
    ax_1.set_ylabel('Closing Price')
    ax_1.set_title(title)
    ax_1.plot(df_1.index,df_1[col])
    ax_1.grid(True,color='0.6',dashes=(5,2,1,2))
    ax_1.set_facecolor((0.9,0.9,0.9))
    
```

```python
#loading data
data='https://raw.githubusercontent.com/derekbanas/TimeSeriesAnalysis/main/S%26P.csv'
```

```python
#using function to display chart\
print_chart(data,'S&P Prices','Close')
```

```python
#loading data
df_tem=pd.read_csv('https://raw.githubusercontent.com/derekbanas/TimeSeriesAnalysis/main/PittsburghWeather.csv',index_col='Date',parse_dates=True)
#removing na value
df_tem=df_tem.dropna()
#choosing data whch are greater than 2000
df_3=df_tem[df_tem.index>'2000-01-01']
df_3
#selecting the size of the chart
f_2=plt.figure(figsize=(12,6),dpi=100)
#adding axes
ax_2=f_2.add_axes([0.0,0.0,0.9,0.9])
#setting y lables
ax_2.set_ylabel('High Temperature')
#setting title of the chart
ax_2.set_title('Pittarburgh Weather')
#ploting the chart with max temperature value
df_tem['tmax'].plot(xlim=['1974-12-21','2017-12-30'])
ax_2.set_facecolor
#creating charrt with mean values of tmax 
df_tem.rolling(window=60).mean()['tmax'].plot(xlim=['1974-12-21','2017-12-30'])
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
sns.get_dataset_names()
flights_df=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/flights.csv')
```

```python
flights_df.head()
```

```python
#creating the column with date
flights_df['day']='01'
flights_df['month']=pd.to_datetime(flights_df.month,
                                format='%B').dt.month
```

```python
flights_df
```

```python
flights_df['date']=pd.to_datetime(flights_df[['year','month','day']])
flights_df
```

```python
flights_df.drop('year',axis=1,inplace=True)
flights_df.drop('month',axis=1,inplace=True)
flights_df.drop('day',axis=1,inplace=True)
flights_df.set_index('date',inplace=True)
```

```python
flights_df
```

```python
flights_df.plot()
f_3=plt.figure(figsize=(12,6),dpi=100)
ax_3=f_3.add_axes([0.0,0.0,0.9,0.9])
ax_3.plot(flights_df.index,flights_df['passengers'])
flights_df.rolling(window=12).mean()['passengers'].plot()
```

```python
f_3=plt.figure(figsize=(12,6),dpi=100)
ax_3=f_3.add_axes([0.0,0.0,0.9,0.9])
ax_3.plot(flights_df.index,flights_df['passengers'])



flights_df['EMA 6']=flights_df['passengers'].ewm(span=6).mean()
flights_df['EMA 6'].plot()
```

```python
sd=seasonal_decompose(flights_df['passengers'],model='multiplication')
sd.trend
sd.seasonal
sd.resid
sd.plot()
```
