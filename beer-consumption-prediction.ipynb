```python
#importing library
import numpy as np
import pandas as pd
import re
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.metrics import r2_score
```

```python
#loading the dataset
df=pd.read_csv('../input/beer-consumption-sao-paulo/Consumo_cerveja.csv')
df
```

```python
#getting information about the dataset
df.info()
```

```python
df.isna().sum()
```

```python
#preprocessing
def preprocess_inputs(df):
    # copying the dataset 
    
    df=df.copy()
    #dropping na values from the dataset
    df=df.dropna(axis=0).reset_index(drop=True)
    for column in ['Temperatura Media (C)','Temperatura Minima (C)','Temperatura Maxima (C)','Precipitacao (mm)']:
        df[column]=df[column].apply(lambda x:np.float(re.sub(r',',".",x)))
    #spliting the data into x and y
    df['Data']=pd.to_datetime(df['Data'])
   
    df['Month']=df['Data'].apply(lambda x:x.month)
    df['Day']=df['Data'].apply(lambda x:x.day)
    df=df.drop('Data',axis=1)
    y=df['Consumo de cerveja (litros)']
    x=df.drop("Consumo de cerveja (litros)",axis=1)
    #Scaling the dataset
    scaler=StandardScaler()
    x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
    return x,y
```

```python
x,y=preprocess_inputs(df)
x
```

```python
#checking datatype
x.dtypes
```

```python
y
```

```python
#Training
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=1)
```

```python
models={
    'Linear Regression':LinearRegression(),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso(),
}
```

```python
for model in models.values():
    model.fit(x_train,y_train)
```

```python
for name,model in models.items():
    print(name+"R^2 Score:{:4f}".format(model.score(x_test,y_test)))
```
