```python
#importing all_data to pandas dataframe
all_data=pd.read_csv('all_data.csv')
#showing data
all_data
```

```python
###Task 2 Add Month Column
```

```python
#Adding month to a column 
all_data['Month']=all_data['Order Date'].str[0:2]
#showing the data
all_data.head()
```

```python
#getting nan values
nan_df=all_data[all_data.isna().any(axis=1)]

nan_df
```

```python
#dropping nan 
all_data=all_data.dropna(how='any')
all_data.head()
```

```python
#gettig column number with index number of column
all_data.columns[0]
```

```python
#dropping the column with drop method
all_data.drop(all_data.columns[0], axis=1, inplace=True)
```

```python
all_data.head(2)
```

```python
#converting string to integer with to_numeric method
pd.to_numeric(all_data['Month'],errors='coerce')
```

```python
#showing the data
all_data.head(2)
```

```python
#getting data with not having or value
all_data=all_data[all_data['Month']!='Or']
```

```python
all_data.head(2)
```

```python
#converting string to integer datatype
all_data['Month']=all_data['Month'].astype('int32')
```

```python
all_data.info()
```

```python
all_data['Quantity Ordered']=all_data['Quantity Ordered'].astype('int32')





```

```python
#getting the information about the dataframe
all_data.info()
```

```python
#converting Price Each colum to float datatype
all_data['Price Each']=all_data['Price Each'].astype('float')

```

```python
#getting sales column
all_data['Sales']=all_data['Quantity Ordered']*all_data['Price Each']
```

```python
all_data.head(2)
```

```python
#applying lambda method to get information about city
all_data['City']=all_data['Purchase Address'].apply(lambda x: x.split(',')[1])
all_data.drop('column', axis=1, inplace=True)
all_data
```

```python
#getting information about total sales
all_data['Sales'].sum()
```

```python
#grouping the month and getting sales information in each month
sales_by_month=all_data.groupby('Month').sum()
sales_by_month
```

```python
plt.figure(figsize=(12,6))
plt.bar(sales_by_month.index,sales_by_month['Sales'])

plt.title('Sales over the month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
#getting sales information in respect of city
city=all_data.groupby('City').sum()
city
```

```python
plt.figure(figsize=(12,6))
plt.bar(city.index,city['Sales'])
plt.title('Sales over the month')
plt.xlabel('City')
plt.ylabel('Sales')
plt.show()
```

```python
#converting order date to date time object
all_data['Order Date']=pd.to_datetime(all_data['Order Date'])
```

```python
all_data.info()
```

```python
#adding hour column to data
all_data['hour']=all_data['Order Date'].dt.hour
all_data['Minutes']=all_data['Order Date'].dt.minute
all_data['Second']=all_data['Order Date'].dt.second
```

```python
all_data.head()
```

```python
hours=[hour for hour,df in all_data.groupby("hour")]
hours
```

```python
#using count method to getting number of sales in a hour
sales_by_hour=all_data.groupby('hour').count()
sales_by_hour=sales_by_hour['Sales']
```

```python
#figure size
plt.figure(figsize=(12,8))
#ploting the chart 
plt.plot(hours,sales_by_hour)
#xlabel
plt.xlabel('hour')
#ylabel
plt.ylabel('Sales Number')
#title
plt.title('Number of Sales in Hours')
#adding grid
plt.grid()
#expanding the chart
plt.xticks(hours)
```
