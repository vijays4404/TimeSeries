# 1. Importing the Libraries

```python
#importing numpy library
import numpy as np
#importing pandas library
import pandas as pd
#importing StandardScaler library to scaling the dataset
from sklearn.preprocessing import StandardScaler
#importing train test split library to split the data between train and test set
from sklearn.model_selection import train_test_split
#import machine learning library XGBoost
import xgboost as xgb
#importing metrics score
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

# Loading the Dataset

```python
df=pd.read_csv('/kaggle/input/automobile-dataset/Automobile_data.csv')
#showing the dataset
df
```

# Getting the Preliminary Information about the dataset

```python
df.info()
```

# Cleaning the Dataset with Replace Function

```python
df['symboling'].unique()
```

```python
df['symboling']=df['symboling'].replace({-1:np.NaN})
```

```python
df['normalized-losses'].unique()
```

```python
df['normalized-losses']=df['normalized-losses'].replace({'?':np.NaN})
```

```python
df=df.replace('?',np.NaN)
```

```python
object_numeric=['normalized-losses','bore','stroke','horsepower','peak-rpm','price']
for column in object_numeric:
    df[column]=df[column].astype(np.float)
```

# Checking of Data Type

```python
df.dtypes
```

# Checking for Missing Values in Each Column

```python
df.isna().sum()
```

```python
df['symboling'].dtypes
```

# Filling Missing Values with Mean of that column

```python
for column in df.columns:
    if df[column].dtypes!='object':
        df[column]=df[column].fillna(df[column].mean())
```

# Now Checking the Missing Values

```python
df.isna().sum()
```

# # ****Filling Missing Value in num-of-doors column with mode of what column

```python
df['num-of-doors']=df['num-of-doors'].fillna(df['num-of-doors'].mode()[0])
```

```python
df['num-of-doors'].unique()
```

# Getting Unique Values for each object columns

```python
{column:list(df[column].unique()) for column in df.columns if df[column].dtypes=='object' }
```

# Replacing object value with numercial value

```python
num_ordering={'two':2,'three':3,'four':4,'five':5,'six':6,'eight':8,'twelve':12}
df['num-of-cylinders']=df['num-of-cylinders'].replace(num_ordering)
```

```python
df['fuel-type']=df['fuel-type'].apply(lambda x:1 if x=='gas' else 0)
```

```python
df['aspiration']=df['aspiration'].apply(lambda x:1 if x=='std' else 0)
```

```python
df['num-of-doors']=df['num-of-doors'].apply(lambda x:2 if x=='two' else x)
```

```python
df['num-of-doors']=df['num-of-doors'].apply(lambda x:4 if x=='four' else x)
```

```python
df['num-of-doors'].dtypes
```

```python
df['engine-location']=df['engine-location'].apply(lambda x:0 if x=='front' else 1)
df['engine-location'].unique() 
```

```python
df
```

```python
one_hot_columns=['make','body-style','drive-wheels','engine-type','fuel-system']
```

# Creating onehot function

```python
def onehot_encode(df,columns):
    for column in columns:
        dummies=pd.get_dummies(df[column])
        df=pd.concat([df,dummies],axis=1)
        df=df.drop(column,axis=1)
    
    return df
```

```python
df=onehot_encode(df,one_hot_columns)
```

```python
df.info()
```

# Splitting the dataset between target and  feature columns

```python
y=df['price']
x=df.drop('price',axis=1)
```

# Scaling the Dataset with StandardScaler

```python
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

# Train Test Split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Train Test Split

# Training the Model

```python
linear_model=LinearRegression()
linear_model.fit(x_train,y_train)
linear_model.score(x_test,y_test)
```

# XG Boost Model

```python
x_train,x_val,y_train,y_val=train_test_split(x_train,y_train,train_size=0.8)
```

```python
d_train=xgb.DMatrix(x_train,label=y_train)
dval=xgb.DMatrix(x_val,label=y_val)
dtest=xgb.DMatrix(x_test,label=y_test)
```

```python
params={'learning_rate':0.001,'max_depth':6}
model=xgb.train(params,d_train,evals=[(dval,'eval')],num_boost_round=2000,early_stopping_rounds=10)
```

```python
y_true=np.array(y_test,dtype=np.float)
y_pred=np.array(model.predict(dtest),dtype=np.float)
```

# R2 Score

```python
r2_score(y_true,y_pred)
```
