```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

# Trying the Predict the Loan_Status from the provided information

```python
#loadingg the data
df=pd.read_csv('/kaggle/input/loan-status-prediction/loan_data.csv')
#showing the dataset
df.head(5)
```

```python
#getting the preliminary information about the dataset
df.info()
```

```python
#dropping loan_id column
df=df.drop('Loan_ID',axis=1)
```

```python
{column:list(df[column].unique()) for column in df.columns if df[column].dtype=='object'}
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
{column:list(df[column].unique()) for column in df.columns if df[column].dtype=='object'}
```

```python
df=df.dropna(axis=0)
```

```python
df['Gender']=df['Gender'].replace({"Male":1,'Female':0})
df['Married']=df['Married'].replace({'Yes':1,'No':0})
df['Dependents']=df['Dependents'].replace({'3+':3}).astype(int)
df['Education']=df['Education'].replace({'Graduate':1,'Not Graduate':0})
df['Self_Employed']=df['Self_Employed'].replace({'Yes':1,'No':0})
df['Property_Area']=df['Property_Area'].replace({'Rural':0,'Semiurban':1,'Urban':2})
df['Loan_Status']=df['Loan_Status'].replace({'Y':1,'N':0})
```

```python
y=df['Loan_Status']
x=df.drop('Loan_Status',axis=1)
```

```python
#scaling the dataset
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns,index=x.index)
```

```python
#train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=1234)
```

```python
#training the model
model=RandomForestClassifier()
model.fit(x_train,y_train)
print('The model score is',model.score(x_test,y_test))
```

```python
model.predict(x_test)
```
