```python
#importing the library
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
```

```python
#loading the dataset
df=pd.read_csv('../input/tictactoe-endgame-dataset-uci/tic-tac-toe-endgame.csv')
#showing the dataset
df
```

```python
#getting unique  value in the column
df['V10'].unique()
```

# Preprocessing

```python
def onehot_encode(df,columns):
    #copying the dataset to a new variable
    df=df.copy()
    #loop through columns
    for column in columns:
        #creating dummies column and savving it to a vvariable
        dummies=pd.get_dummies(df[column],prefix=column)
        #concating the original dataset to a dummies dataset side by side
        df=pd.concat([df,dummies],axis=1)
        #droping the original column
        df=df.drop(column,axis=1)
    #returning the dataframe
    return df
        
        
    
```

```python
#Preprocess Inputs
def preprocess_inputs(df):
    #copying the dataframe
    df=df.copy()
    #Encode label values as numbers
    df['V10']=df['V10'].replace({'negative':0,'positive':1})
    #calling the onehot_encode function
    df=onehot_encode(
    df,columns=['V' + str(i) for i in range(1,10)]
    )
    #Split df into x and y
    y=df['V10'].copy()
    x=df.drop('V10',axis=1).copy()
    #train_test_split
    x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=1)
    return x_train,x_test,y_train,y_test
```

```python
x_train,x_test,y_train,y_test=preprocess_inputs(df)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Training

```python
#creating the dictionary of the model
models={
    'K-Nearest Neighbors':KNeighborsClassifier(),
    'Logistic Regression': LogisticRegression(),
    'Support Vector Machine': SVC(),
    'Decision Tree': DecisionTreeClassifier(),
    'Neural Network': MLPClassifier()
}
```

```python
models
```

```python
for name,model in models.items():
    model.fit(x_train,y_train)
    print(name+'trained')
```

# Result

```python
#looping through the list of models.
for name, model in models.items():
    print(name+'Accuracy:{:2f}%'.format(model.score(x_test,y_test)*100))
```
