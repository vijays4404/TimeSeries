```python
#importing library
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
#loading the dataset
df=pd.read_csv('../input/real-estate-dataset/data.csv')
df
```

```python
#getting information about the dataset
df.info()
```

```python
#checking for missing values
df['RM'].isna().sum()
```

```python
#filling missing values with mean of the column
df['RM']=df['RM'].fillna(df['RM'].mean())
```

```python
#splitting the data between x and y
y=df['MEDV']
x=df.drop('MEDV',axis=1)
```

```python
#scaling the dataset
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
x
```

```python
#train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
#getting shape of the model
print(x_train.shape)
print(x_test.shape)
```

```python
bootstrap_model=RandomForestRegressor(n_estimators=100,criterion='mse',bootstrap=True)
bootstrap_model.fit(x_train,y_train)
```

```python
#printing the result
print(bootstrap_model.score(x_test,y_test))
```

```python
#no bootstrap
no_bootstrap_model=RandomForestRegressor(n_estimators=100,criterion='mse',bootstrap=False)
no_bootstrap_model.fit(x_train,y_train)
```

```python
print(no_bootstrap_model.score(x_test,y_test))
```

```python
random=RandomForestRegressor()
random.fit(x_train,y_train)
```

```python
print(random.score(x_test,y_test))
```
