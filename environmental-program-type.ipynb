```python
#importing library
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
pd.set_option('max_columns',None)
```

```python
#loading the dataset
df=pd.read_csv('../input/nys-environmental-remediation-sites/environmental-remediation-sites.csv')
df
```

```python
#getting the information about the dataset
df.info()
```

```python
#checking misisng values in each column
df.isna().sum()
```

```python
#column more than 25 % null values 
null_column=df.loc[:,df.isna().sum()>0.25*df.shape[0]]
```

```python
#dropping the unneeded column
df=df.drop(null_column,axis=1)
```

```python
#again checking the null values in each column
df.isna().sum()
```

```python
#column to drop
unneeded_columns=['Program Number','Project Name',
               'Program Facility Name','Address1',
               'Locality','ZIPCode','SWIS Code',
               'Owner Name','Owner Address1',
               'Owner City','Owner State','Owner ZIP'
               ]
```

```python
#uneeded column
df.drop(unneeded_columns,axis=1,inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
df=df.drop(['Contaminants',"Georeference"],axis=1)
```

```python
#defining a function to get unique values in a column
def get_unique(df,columns):
    return {column:list(df[column].unique()) for column in columns}
def get_categorical_column(df):
    return {column for column in df.columns if df.dtypes[column]=='object'}
```

```python
#checking the unique 
get_unique(df,get_categorical_column(df))
```

```python
df['Project Completion Date']
```

```python
#Extracting month and years from Project Completion Data Column
df['Project Completion Date'].isnull().sum()
```

```python
df['Project Completion Date']=df['Project Completion Date'].apply(lambda x:x[0:7] if str(x)!='nan' else x)
```

```python
#Extracting Year and Month from the Project Completion column
df['Year']=df['Project Completion Date'].apply(lambda x:np.float(x[0:4]) if str(x)!='nan' else x)
df['Month']=df['Project Completion Date'].apply(lambda x:np.float(x[5:7]) if str(x)!='nan' else x)
#Droping the project completion column
df=df.drop('Project Completion Date',axis=1)
```

```python
df
```

```python
#Again checking the missing values
df.isna().sum()
```

```python
for column in ['New York Zip Codes 2','Counties 2','Year','Month']:
    df[column]=df[column].fillna(df[column].mean())
```

```python
#Again checking the missing values
df.isna().sum()
```

```python
def onehot_encode(df,column):
    df=df.copy()
    dummies=pd.get_dummies(df[column])
    df=pd.concat([df,dummies],axis=1)
    df=df.drop(column,axis=1)
    return df
```

```python
nominal_features=get_categorical_column(df)
nominal_features.remove('Program Type')
```

```python
for feature in nominal_features:
    df=onehot_encode(df,feature)
    
```

```python
df
```

```python
(df.dtypes=='object').sum()
```

```python
label_encoder=LabelEncoder()
df['Program Type']=label_encoder.fit_transform(df['Program Type'])
```

```python
#checkng mapping for label_encoder
target_mapping={index:column for index,column in enumerate(label_encoder.classes_)}
```

```python
target_mapping
```

```python
#Splitng and Scaling
y=df['Program Type']
x=df.drop('Program Type',axis=1)
```

```python
#scaling the feature dataset
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),index=x.index,columns=x.columns)
```

```python
#checking the shape of x
x.shape
```

```python
#spliting the data
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=1)
```

```python
#setting up the model
inputs=tf.keras.Input(shape=(118,))
x=tf.keras.layers.Dense(64,activation='relu')(inputs)
x=tf.keras.layers.Dense(64,activation='relu')(x)
outputs=tf.keras.layers.Dense(5,activation='softmax')(x)
model=tf.keras.Model(inputs=inputs,outputs=outputs)
```

```python
model.summary()
```

```python
#compilng the model
model.compile(optimizer='adam',
                 loss='sparse_categorical_crossentropy',
                 metrics=['accuracy'])


batch_size=64
epochs=60
history=model.fit(
    x_train,
    y_train,
    validation_split=0.2,
    batch_size=batch_size,
    epochs=epochs,
    callbacks=[tf.keras.callbacks.ReduceLROnPlateau()],
    verbose=0)
```

```python
#ploting the 
fig=px.line(
    history.history,y=['loss','val_loss'],
    labels={'index':'Epoch','value':'Loss'},
    title='Training and Validation Loss')
fig.show()
```

```python
#Evaluating model
model.evaluate(x_test,y_test)
```
