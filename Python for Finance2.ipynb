```python
#Daily Returns for DataFrame
# We calculate a percetnage rate of return for eack day to compare investment.
#Simple Rate of Return =(End Price-Beginning Price)/Beginning Price or
# (EP/BP)-1
def add_daily_return_to_df(df,ticker):
    df['daily-return']=(df['Adj Close']/df['Adj Close'].shift(1))-1
    df.to_csv(path+'.csv')
    return df
```

```python
#Returns return on investment over time
#This amount does not inlude your initial investment
#ROI(Rate of Investment)=End Value(200)-Initial Value(100)/Initial Value=1
#Your new total is initial invesment+1*Initial Invesment=200
def get_roi_defined_time(df):
    df['Date']=pd.to_datetime(df['Date'])
    start_val=df[df['Date']==s_date_str]['Adj Close'][0]
    end_val=df[df['Date']==e_date_str]['Adj Close'][0]
    print('Initial Price:',start_val)
    print('Final Price',end_val)
    roi=(end_val-start_val)/start_val
    return roi
```

```python
#Get Coefficient of Variation
def get_cov(stock_df):
    mean=stock_df['Adj Close'].mean()
    sd=stock_df['Adj Close'].std()
    cov=sd/mean
    return cov
```
