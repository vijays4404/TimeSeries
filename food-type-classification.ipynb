```python
#importing library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,LabelEncoder
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
#loading the dataset
df=pd.read_csv('/kaggle/input/indian-food-101/indian_food.csv')
#showing the dataset
df
```

```python
#checking for null values in the dataset
df.isna().sum()
```

```python
food_vocab=set()

for ingredient in df['ingredients']:
    for food in ingredient.split(','):
        if food.strip().lower() not in food_vocab:
            food_vocab.add(food.strip().lower())

food_vocab
    
```

```python
food_columns=pd.DataFrame()


for i,ingredient in enumerate(df['ingredients']):
    for food in ingredient.split(','):
        if food.strip().lower() in food_vocab:
            food_columns.loc[i,food.strip().lower()]=1
food_columns.fillna(0)
```

```python
df.drop(list(df.columns[0:2]),axis=1,inplace=True)
```

```python
df
```

```python
{column:list(df[column].unique()) for column in df.columns if df[column].dtypes=='object'}
```

```python
df=df.replace({'-1': np.NaN})
```

```python
{column:list(df[column].unique()) for column in df.columns if df[column].dtypes=='object'}
```

```python
def onehot_encode(df,columns,prefixes):
    df=df.copy()
    for column,prefix in zip(columns,prefixes):
        dummies=pd.get_dummies(df[column],prefix=prefix)
        df=pd.concat([df,dummies],axis=1)
        df=df.drop(column,axis=1)
    return df
        
```

```python
onehot_columns=['flavor_profile','course','state','region']
prefixes=['fp','cour','st','reg']
df=onehot_encode(df,columns=onehot_columns,prefixes=prefixes)
```

```python
for column in df.columns:
    if df[column].dtypes=='bool':
        df[column]=df[column].astype(int)
        
        
```

```python
df=df.replace({-1:np.NaN})
```

```python
df.isna().sum()
```

```python
for column in ['prep_time','cook_time']:

    df[column]=df[column].fillna(df[column].mean())
```

```python
df
```

```python
df.dtypes
```

```python
label=LabelEncoder()

df['diet']=label.fit_transform(df['diet'])

```

```python
{index:label for index,label in enumerate(label.classes_)}
```

```python
#Spliting and scaling data
y=df['diet']
x=df.drop('diet',axis=1)
scaler=StandardScaler()

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=123)

scaler.fit(x_train)
x_train=pd.DataFrame(scaler.transform(x_train),index=x_train.index,columns=x_train.columns)
x_test=pd.DataFrame(scaler.transform(x_test),index=x_test.index,columns=x_test.columns)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
