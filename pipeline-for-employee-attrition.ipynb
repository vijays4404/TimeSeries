```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder,StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('/kaggle/input/employee-attrition-data/MFG10YearTerminationData.csv')
#showing the dataset
df
```

```python
def preprocess_inputs(df):
    #creating the copy of the dataset
    df=df.copy()
    #dropping the unnecessary columns
    df=df.drop(['EmployeeID','gender_short'],axis=1)
    df=df.drop(['terminationdate_key','length_of_service','termreason_desc','termtype_desc'],axis=1)
    
    
    
    
    y=df['STATUS']
    x=df.drop('STATUS',axis=1)
    
    x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
    
    
    return x_train,x_test,y_train,y_test
```

```python
x_train,x_test,y_train,y_test=preprocess_inputs(df)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
binary_features=['gender_short','gender_full','STATUS','BUSINESS_UNIT']
nominal_features=['city_name','department_name','job_title']
date_features=['recorddate_key','birthdate_key','orighiredate_key']


binary_transformer=Pipeline(steps=[('Ordinal',OrdinalEncoder(categories='auto'))])

nominal_transformer=Pipeline(steps=[('Nominal',OneHotEncoder())])




date_transformer=Pipeline(steps=[('Date',DateTransformer())])





```

```python
preprocessor=ColumnTransformer(transformers=[('binary',binary_transformer,binary_features),('nominal',nominal_transformer,nominal_features),
                                             ('date',date_transformer,date_features)])
```

```python
preprocessor
```

```python
#Building the Final Pipeline
model=Pipeline(steps=[('preprocessor',preprocessor),('scaler',StandardScaler()),('classifier',LogisticRegression())])
```

```python
class DateTransformer:
    
    def fit(self,x,y):
        return self
        
    def transform(self,x):
        
        x[column]=pd.to_datetime(x[column])
        
        x[column+'_year']=x[column].dt.year
        x[column+'_month']=x[column].dt.month
        x[column+'_day']=x[column].dt.day
        x=x.drop(column,axis=1)
        return x
```
