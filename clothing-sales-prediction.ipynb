```python
#storing the size in a list and assigning it to a variable
size_ordering=['XXS','XS','S','M','L','XL','XXL']
size_ordering
```

```python
#creating a function with ordinal encode the columns with orders
def ordinal_encode(df,column,ordering):
    return df[column].apply(lambda x:ordering.index(x) if x in ordering else None)
```

```python
#applying the function to a variable
df['product_variation_size_id']=ordinal_encode(df,'product_variation_size_id',size_ordering)
```

```python
df['product_variation_size_id']
```

```python
df['product_color']
```

```python
#creating one hot encode function
def onehot_encode(df,column):

    #creating the dummies column with pandas get_dummies function
    dummies=pd.get_dummies(df[column])
    #concating the dummies columes with original dataset
    df=pd.concat([df,dummies],axis=1)
    #dropping the orginal column
    df=df.drop(column,axis=1)
    #returning the dataframe
    return df
```

```python
df=onehot_encode(df,'product_color')
df
```

```python
#Checking the missing values
df.isna().sum()
```

```python
null_columns=['rating_five_count','rating_four_count','rating_three_count',
'rating_two_count','rating_one_count',
'product_variation_size_id']
```

```python
null_columns
```

```python
for column in null_columns:
    df[column]=df[column].fillna(df[column].mean())
```

```python
df.isna().sum()
```

```python
#spliting between target and feature set

y=df['units_sold']
x=df.drop('units_sold',axis=1)
```

```python
#scaling the data with MinMaxScaler
scaler=MinMaxScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

```python
y.unique()
```

```python
encoder=LabelEncoder()
y=encoder.fit_transform(y)
y_mappings={index:label for index,label in enumerate(encoder.classes_)}
```

```python
#Training the dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
#tensorflow
inputs=tf.keras.Input(shape=(124,))
x=tf.keras.layers.Dense(16,activation='relu')(inputs)
x=tf.keras.layers.Dense(16,activation='relu')(x)
outputs=tf.keras.layers.Dense(15,activation='softmax')(x)
```

```python
model=tf.keras.Model(inputs=inputs,outputs=outputs)
```

```python
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
batch_size=32
epochs=300
history=model.fit(x_train,y_train,validation_split=0.2,batch_size=batch_size,epochs=epochs,
         verbose=0)
```

```python
plt.figure(figsize=(14,19))
epochs_range=range(1,epochs+1)
train_loss=history.history['loss']
val_loss=history.history['val_loss']
plt.plot(epochs_range,train_loss,label='Training Loss')
plt.plot(epochs_range,val_loss,label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
model.evaluate(x_test,y_test)
```
