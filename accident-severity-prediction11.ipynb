```python
#trying to predict the severity of accident

#importing the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import tensorflow as tf

```

```python
#loading the dataset
df=pd.read_csv('/kaggle/input/us-accidents/US_Accidents_March23.csv',nrows=400000)
```

```python
df
```

```python
#getting preliminary information about the dataset

df.info()
```

```python
#checking for the missing value in the dataset
null_columns=['End_Lat','End_Lng','Wind_Chill(F)','Precipitation(in)']
#dropping the columns with high null values

df=df.drop(null_columns,axis=1)
```

```python
df.isna().sum()
```

```python
df=df.dropna(axis=0).reset_index(drop=True)
```

```python
df.isna().sum()
```

```python
#dropping unnecesary columns 

{column:len(df[column].unique()) for column in df.columns if df[column].dtypes=='object'}
```

```python
unneeded_columns=['ID','Description','Street','Zipcode','City','Country']
df=df.drop(unneeded_columns,axis=1)
```

```python
df
```

```python
time_stamp=['Start_Time','End_Time','Weather_Timestamp']

for column in time_stamp:
    df[column+'_year']=pd.to_datetime(df[column]).dt.year
    df[column+'_month']=pd.to_datetime(df[column]).dt.month
    df[column+'_day']=pd.to_datetime(df[column]).dt.day
    df[column+'_hour']=pd.to_datetime(df[column]).dt.hour
    df[column+'_minute']=pd.to_datetime(df[column]).dt.minute
    df[column+'_second']=pd.to_datetime(df[column]).dt.second
```

```python
df=df.drop(time_stamp,axis=1)
```

```python
df
```

```python
def onehot_encode(df,columns):
    df=df.copy()
    for column in columns:
        dummies=pd.get_dummies(df[column],prefix=column)
        df=pd.concat([df,dummies],axis=1)
        df=df.drop(column,axis=1)
    return df
```

```python
onehot_columns=['County','State','Timezone','Airport_Code','Wind_Direction','Weather_Condition']


df=onehot_encode(df,onehot_columns)
```

```python
df['Source'].value_counts()
```

```python
df
```

```python
for column in [column for column in df.columns if df[column].dtypes=='bool']:
    df[column]=df[column].astype(np.int)
```

```python
df
```

```python
df['Source']=df['Source'].apply(lambda x:1 if x=='Source1' else 0)
```

```python
df
```

```python
for column in [column for column in df.columns if df[column].dtypes=='object']:
    df[column]=df[column].apply(lambda x:1 if x=='Day' else 0)
```

```python
[column for column in df.columns if df[column].dtypes=='object']
```

```python
y=df['Severity']
x=df.drop('Severity',axis=1)
```

```python
y=y-1
```

```python
y.value_counts()
```

```python
#train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=123)
```

```python
#scaling 
scaler=StandardScaler()
scaler.fit(x_train)
x_train=pd.DataFrame(scaler.transform(x_train),columns=x_train.columns)
x_test=pd.DataFrame(scaler.transform(x_test),columns=x_test.columns)
```

```python
x_train
```

```python
print(x_train.shape)
```

```python
#contructing the tensorflow model


inputs=tf.keras.Input(shape=(1352,))
x=tf.keras.layers.Dense(64,activation='relu')(inputs)
x=tf.keras.layers.Dense(64,activation='relu')(x)
outputs=tf.keras.layers.Dense(4,activation='softmax')(x)



model=tf.keras.Model(inputs,outputs)
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])


batch_size=32
epochs=20





history=model.fit(x_train,y_train,validation_split=0.2,batch_size=batch_size,epochs=epochs,
                 callbacks=[tf.keras.callbacks.ReduceLROnPlateau(),
                           tf.keras.callbacks.EarlyStopping(monitor='val_loss',patience=3,restore_best_weights=True)])
```

```python
print("Model's Accuracy",model.evaluate(x_test,y_test))
```
