# Preprocessing Function

```python
def preprocess_inputs(df):
    #creating copy of the dataset
    df=df.copy()
    #dropping bad rows
    bad_rows=df.query("crawlStatus=='unavailable'").index
    #dropping the bad_rows from the dataset
    df=df.drop(bad_rows,axis=0)
    
    
    
    
    #returning the dataset
    #selecting only useful column
    df=df[['areaSqm','city','furnish','latitude','propertyType',
          'rent','internet','kitchen','living','pets','shower','smokingInside','toilet']]
    #replacing missing and unknown with np.NaN
    df=df.replace({'':np.NaN,'Unknown':np.NaN})
    #replacing na with mode of that column
    
    for column in df.columns:
        df[column]=df[column].fillna(df[column].mode()[0])
    
    #splitting the data between x and y
    y=df['rent']
    x=df.drop('rent',axis=1)
    #train_test_split
    x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=1,shuffle=True)
    
    
    
    return x_train,x_test,y_train,y_test
```

```python
x_train,x_test,y_train,y_test=preprocess_inputs(df)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Building PipeLine and Training

```python
#tree models don't require scaling
binary_columns=['internet']
binary_transformation=Pipeline(steps=[('ordinal',OrdinalEncoder())])
onehot_transformation=Pipeline(steps=[('onehot',OneHotEncoder(sparse=False,handle_unknown='ignore'))])

preprocessing=ColumnTransformer(transformers=[('binary',binary_transformation,binary_columns),('onehot',onehot_transformation,ordinal_columns)],
                               remainder='passthrough')


model=Pipeline(steps=[('preprocesing',preprocessing),('regressor',RandomForestRegressor())])
```

# Training the Model with Train Set

```python
model.fit(x_train,y_train)
```
