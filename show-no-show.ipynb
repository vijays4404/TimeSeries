# Cleaning

```python
#checking for missing values
print('Total Missing Value',df.isna().sum().sum())
```

```python
#checking for number of unique values
len(df['PatientId'].unique())
```

```python
#dropping patient id from the dataset
df=df.drop(['PatientId','AppointmentID'],axis=1)
```

```python
#renaming the column name
df=df.rename(columns={'ScheduledDay':'ScheduleDate','AppointmentDay':'AppointmentDate'})
```

```python
df
```

```python
df['ScheduleYear']=df['ScheduleDate'].apply(lambda x:np.int(x[0:4]))
df['ScheduleMonth']=df['ScheduleDate'].apply(lambda x:np.int(x[5:7]))
df['ScheduleDay']=df['ScheduleDate'].apply(lambda x:np.int(x[8:10]))
df['ScheduleHour']=df['ScheduleDate'].apply(lambda x:np.int(x[11:13]))
df['ScheduleMinute']=df['ScheduleDate'].apply(lambda x:np.int(x[14:16]))
df['ScheduleSecond']=df['ScheduleDate'].apply(lambda x:np.int(x[17:19]))
df=df.drop('ScheduleDate',axis=1)
```

```python
df
```

```python
df['AppointmentYear']=df['AppointmentDate'].apply(lambda x:np.int(x[0:4]))
df['AppointmentMonth']=df['AppointmentDate'].apply(lambda x:np.int(x[5:7]))
df['AppointmentDay']=df['AppointmentDate'].apply(lambda x:np.int(x[8:10]))


df=df.drop('AppointmentDate',axis=1)
```

```python
df
```

# Encoding

```python
#selecing only object columns
{column:list(df[column].unique()) for column in df.select_dtypes('object').columns}
```

```python
#creating function for encoding
def binary_encode(df,column,positive_value):
    df=df.copy()
    df[column]=df[column].apply(lambda x:1 if x==positive_value else 0)
    return df
```

```python
def onehot_encode(df,column,prefix):
    df=df.copy()
    dummies=pd.get_dummies(df[column],prefix=prefix)
    df=pd.concat([df,dummies],axis=1)
    df=df.drop(column,axis=1)
    return df
```

```python
df=binary_encode(df,'Gender','M')
df=binary_encode(df,'No-show','Yes')
```

```python
df=onehot_encode(df,'Neighbourhood','N')
```

```python
df
```

# Spliting and Scaling the Dataset

```python
y=df['No-show']
x=df.drop('No-show',axis=1)
```

# Scaling the Dataset with StandardScaler

```python
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

# train test 

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

x_train.shape

```python
x_train.shape
```

# Training with Tensorflow

```python
inputs=tf.keras.Input(shape=(98,))
x=tf.keras.layers.Dense(64,activation='relu')(inputs)
x=tf.keras.layers.Dense(64,activation='relu')(x)
outputs=tf.keras.layers.Dense(1,activation='sigmoid')(x)
```

# Creating the Model

```python
model=tf.keras.Model(inputs,outputs)
```

# Compiling the model

```python
model.compile(optimizer='adam',loss='binary_crossentropy',
             metrics=['accuracy',tf.keras.metrics.AUC(name='auc')])
```

# Fitting the Model

```python
history=model.fit(x_train,y_train,validation_split=0.2,batch_size=32,
                 epochs=100,callbacks=[tf.keras.callbacks.EarlyStopping(
                 monitor='val_loss',patience=3,restore_best_weights=True)])
```

# Result

```python
model.evaluate(x_test,y_test)
```
