```python
#importing library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
#loading data
df=pd.read_csv('https://raw.githubusercontent.com/sahidul-shaikh/car-price-prediction-linear-regression/main/CarPrice_Assignment.csv')
df.head(5)
```

```python
len(df.info())
```

```python
#checking the number of missing values
len(df.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#encoding the data
for column in df.columns:
    if df[column].dtype==np.number:
        continue
    df[column]=LabelEncoder().fit_transform(df[column])
```

```python
df
```

```python
#creating x and y variable
x=df.drop('price',axis=1)#feature data dataset
x=x.drop('car_ID',axis=1)
#target column
y=df['price']
```

```python
#spliting the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#Create the model
model=LinearRegression()
#Train the model
model.fit(x_train,y_train)
```

```python
#Create the predictions on the test data
prediction=model.predict(x_test)
#print the prediction
print(prediction)
prediction_train=model.predict(x_train)
print(prediction_train)
```

```python
#Show the models score on the training and test data sets
error_score=metrics.r2_score(y_test,prediction)
print(error_score)
```

```python
error_score_train=metrics.r2_score(y_train,prediction_train)
print(error_score_train)
```
