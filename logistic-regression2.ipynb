```python
#importing library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
#loading the dataset
df=pd.read_csv('../input/the-ultimate-halloween-candy-power-ranking/candy-data.csv')
#showing the dataset
df
```

```python

names=df['competitorname']
#dropping the competitor name column from the dataset
df.drop('competitorname',axis=1,inplace=True)
```

# Visualization

```python
#plotting the corelation matrix
plt.figure(figsize=(12,18))
sns.heatmap(df.corr(),annot=True,vmin=-1,vmax=1)
plt.show()
```

```python
#checking for null values
df.isnull().sum()
```

```python
#Splitting the data between x and y
y=df['bar']
x=df.drop('bar',axis=1)
```

```python
#Scaling
scaler=MinMaxScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

```python
x
```

```python
#train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=1)
```

```python
#training the dataset
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
print(f'Model Accuracy{model.score(x_test,y_test)}')
```
