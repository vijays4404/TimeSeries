# Preprocessing the Dataset

```python
#columns to drops
df=df.drop(df.columns[5:9],axis=1)
```

```python
df=df.drop('Publisher',axis=1)
```

```python
#checking for null values
df.isna().sum()
```

```python
#immputing null values
df['Year_of_Release']=df['Year_of_Release'].fillna(df['Year_of_Release'].mean())
```

```python
developer_counts=df['Developer'].value_counts()
df['Developer']=df['Developer'].apply(lambda x:'Small Developer' if developer_counts[x]<50 else x)
```

```python
def onehot_encoding(df,columns):
    df=df.copy()
    for column in columns:
        dummies=pd.get_dummies(df[column],dtype='bool')

        df=pd.concat([df,dummies],axis=1)
        df=df.drop(column,axis=1)
    return df
```

```python
df=df.drop('Name',axis=1)
```

```python
categorical_columns=[column for column in df.columns if df[column].dtypes=='object']
```

```python
for column in df.columns:
    if df[column].dtypes=='bool':
        df[column]=df[column].apply(lambda x:1 if x=='True' else 0)
```

```python
y=df['Global_Sales']
x=df.drop('Global_Sales',axis=1)
```

```python
#xtrain_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=123,shuffle=True)
```

```python

scaler=StandardScaler()
scaler.fit(x_train)
x_train=pd.DataFrame(scaler.transform(x_train),columns=x_train.columns,index=x_train.index)
x_test=pd.DataFrame(scaler.transform(x_test),columns=x_test.columns,index=x_test.index)
```

```python
y_train
```

# Training the Model

```python
model=RandomForestRegressor()
model.fit(x_train,y_train)
print(model.score(x_test,y_test))
```
