```python
#importing libarary
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
import seaborn as sns
```

```python
beauty=pd.read_csv('TeachingRatings.csv')
```

```python
beauty.describe()
```

```python
#data for  y 
y=beauty['beauty']
#data for x
xl=beauty['eval']
# plotting scattered plot
plt.scatter(xl,y)
#labeling x and y axis
plt.xlabel('Evaluation')
plt.ylabel('Beauty Rating')
#incept where the x hit the y axis
x=sm.add_constant(xl)
result=sm.OLS(y,x).fit()
#summary of this data
result.summary()
yhat=0.2687*xl-1.0743
fig=plt.plot(xl,yhat,lw=4,c='orange',label='Regression')
```

```python
sns.pairplot(beauty)
```
