```python
#importing the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Creating empty list
Year=[]
Yearly_Income=[]
Yearly_Expense=[]
Annual_Return=[]
Yearly_Investments=[]
```

```python
#yearly income
income=35000
#yearly expense
expense=income/2
#yearly interest
interest_rate=0.08
#You invest half of your income
investment=income/2
#calcuating the annual return
annual_return=investment*interest_rate
#getting the current year
year=2021
```

```python
#Append the first vales to the list
Year.append(year)
Yearly_Income.append(income)
Yearly_Expense.append(expense)
Yearly_Investments.append(investment)
Annual_Return.append(annual_return)
```

```python
#loop for 'n' years
invested_years=20
for i in range(0,invested_years-1):
    investment =investment+annual_return+income/2
    annual_return=investment*interest_rate
    Year.append(year+i+1)
    Yearly_Income.append(income)
    Yearly_Expense.append(expense)
    Annual_Return.append(annual_return)
    Yearly_Investments.append(investment)
    
```

```python
#Creating a dataframe
df=pd.DataFrame()
#Adding columns to dataframe
df['Year']=Year
df['Yearly_Income']=Yearly_Income
df['Yearly_Expenses']=Yearly_Expense
df['Yearly_Investments']=Yearly_Investments
df['Annual_Returns']=Annual_Return
df
```

```python
#visually showing the data
plt.figure(figsize=(12,5))
plt.plot(df['Year'],df['Yearly_Expenses'],label='Yearly_Expenses')
plt.plot(df['Year'],df['Annual_Returns'],label='Annual_Returns')
plt.xticks(rotation=45)
plt.title('Retire in 10 years')
plt.xlabel('Year')
plt.ylabel('$')
plt.legend()
plt.show()
```
