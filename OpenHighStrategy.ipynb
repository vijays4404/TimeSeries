```python
#importing library
import pandas as pd
```

```python
#loading the data
df=pd.read_csv('StrategyReports_AAPL_1322.csv')
df

```

```python
#adding temporary column
columns=['Script','Date','Name']
df.columns=columns
```

```python
#showing the data
df
```

```python
#using split method to split the data
first=df['Script'].str.split(';',expand=True)
first
```

```python
#adding column to the data
columns=['Sno','Script','Buy/Sell','Stock','Price','Date']
first.columns=columns
first
```

```python
#spliting second column
second=df['Date'].str.split(';',expand=True)
columns=['Time','Change1','Change2','Change3','Change4']
second.columns=columns
second
```

```python
#combining after spliting 
df = pd.concat([first,second], axis=1)
df
```

```python
#getting information about data
df.info()
```

```python
#converting to data_time
df['Date']=pd.to_datetime(df['Date'])
```

```python
#setting index 
df.set_index('Time', inplace=True)
df
```

```python
#getting data for specific timeframe
match_timestamp = "09:30:00"
first=df.loc[df.index.strftime("%H:%M:%S") == match_timestamp]
nine=pd.DataFrame(first['Price'])
nine
nine = nine[~nine.index.duplicated(keep='first')]
nine


nine.rename(columns = {'Price':'Opening'}, inplace = True)
nine
```

```python
#getting data for specific timeframe
match_timestamp = "11:00:00"
second=df.loc[df.index.strftime("%H:%M:%S") == match_timestamp]
eleven=pd.DataFrame(second['Price'])
eleven
eleven = eleven[~eleven.index.duplicated(keep='first')]
eleven
eleven.rename(columns = {'Price':'Closing'}, inplace = True)
eleven
```

```python
#removing time from the data
nine.set_index(nine.index.date,inplace=True)
nine
```

```python
#removing time from index
eleven.set_index(eleven.index.date,inplace=True)
eleven
```

```python
#combining both dataframe
final = pd.concat([nine['Opening'], eleven['Closing']], axis=1)
final
```

```python
#shifting period

final['Closing Prev']=final['Closing'].shift(periods=1)
final
```

```python
#Adding new column
final['High']=final['Closing']>final['Closing Prev']
```
