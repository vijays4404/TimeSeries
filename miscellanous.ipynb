```python
#importing the library
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings(action='ignore')
```

```python
#loading the dataset
train_df=pd.read_csv('/kaggle/input/aps-failure-at-scania-trucks-data-set/aps_failure_training_set_processed_8bit.csv')
train_labels=pd.read_csv('/kaggle/input/aps-failure-at-scania-trucks-data-set/aps_failure_training_set.csv')['class']

test_df=pd.read_csv('/kaggle/input/aps-failure-at-scania-trucks-data-set/aps_failure_test_set_processed_8bit.csv')
test_labels=pd.read_csv('/kaggle/input/aps-failure-at-scania-trucks-data-set/aps_failure_test_set.csv')['class']

```

```python
train_df
```

```python
train_df['class']=train_labels
```

```python
test_df['class']=test_labels
```

# Preprocessing

```python
y_train=train_df['class'].copy()
x_train=train_df.drop('class',axis=1)
y_test=test_df['class'].copy()
x_test=test_df.drop('class',axis=1)
```

```python
x_train
```

```python
y_train
```

# Training Without dimensionality reduction

```python
model=LogisticRegression()
model.fit(x_train,y_train)
print('Test Accuracy {:.2f}'.format(model.score(x_test,y_test)*100))
```

# Dimensionality Reduction with PCA

```python
n_components=170
pca=PCA(n_components=n_components)
pca.fit(x_train)
x_train_reduced=pd.DataFrame(pca.transform(x_train),columns=['PC'+ str(i+1) for i in range(n_components)],index=x_train.index)
x_test_reduced=pd.DataFrame(pca.transform(x_test),columns=['PC'+str(i+1) for i in range(n_components)],index=x_test.index)
```
