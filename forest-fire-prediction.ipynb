```python
#importing library
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.neural_network import MLPRegressor,MLPClassifier
import warnings
warnings.filterwarnings(action='ignore')
```

```python
#loading the dataset
df=pd.read_csv('/kaggle/input/forest-fires-data-set/forestfires.csv')
df
```

# # Helper Function# 

```python
def ordinal_encode(df,column,ordering):
    df=df.copy()
    df[column]=df[column].apply(lambda x:ordering.index(x) )
    return df
```

# Preprocessing Function

```python
def preprocessing_inputs(df,task):
    df=df.copy()
    #Ordinal Encoding months and day columns

    df=ordinal_encode(df,'month',['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec'])
    df=ordinal_encode(df,'day',['sun','mon','tue','wed','thu','fri','sat'])
    
    if task=='regression':
        y=df['area']
    elif task=='classification':
        y=df['area'].apply(lambda x:1 if x>0 else 0)

    x=df.drop('area',axis=1)

    #train_test_split
    x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=123,shuffle=True)
    scaler=StandardScaler()

    scaler.fit(x_train)
    
    x_train=pd.DataFrame(scaler.transform(x_train),columns=x_train.columns,index=x_train.index)
    x_test=pd.DataFrame(scaler.transform(x_test),columns=x_test.columns,index=x_test.index)

    
    return x_train,x_test,y_train,y_test
```

```python
x_train,x_test,y_train,y_test=preprocessing_inputs(df,'classification')
print(x_train.shape)
```

```python
x_train
```

# # Training the Model# 

```python
log_model=LogisticRegression()
log_model.fit(x_train,y_train)
print('The accuracy of the logistic regression',log_model.score(x_test,y_test))
```

```python
nn_model=MLPClassifier(hidden_layer_sizes=(16,16))
nn_model.fit(x_train,y_train)
print('The accuracy of the model is',nn_model.score(x_test,y_test))
```

```python
x_train,x_test,y_train,y_test=preprocessing_inputs(df,'regression')

```

```python
lin_model=LinearRegression()

lin_model.fit(x_train,y_train)
print('The accuray of linear model',lin_model.score(x_test,y_test))
```

```python
nn_model=MLPRegressor(hidden_layer_sizes=(16,16))
nn_model.fit(x_train,y_train)
print('The accuracy of the model is',nn_model.score(x_test,y_test))
```
