```python
#creating a function
def save_to_csv_from_yahoo(ticker,syear,smonth,sday,eyear,emonth,eday):
    #storing start date in start variable
    start=dt.datetime(syear,smonth,sday)
    #storing end date in end variable
    end=dt.datetime(eyear,emonth,eday)
    #loading stock date in df dataframe
    df=web.DataReader(ticker,'yahoo',start,end)
    #saving data in csv format to particular folder
    df.to_csv("C:\\Users\\ASUS\\Desktop\\db\\"+ticker+'.csv' )
    #returning data
    return df
    
    
```

```python
#Adding Daily Return to Dataframe
#creating a function
def add_daily_return_to_df(df,ticker):
    #creating a new column to dataframe which is daily_return
    df['daily_return']=(df['Adj Close']/df['Adj Close'].shift(1))-1
    #saving the new file
    df.to_csv("C:\\Users\\ASUS\\Desktop\\db\\"+ticker+'.csv')
    #returning the dataframe
    return df
    
```

```python

#creating a function to return total return in specific time
def get_return_defined_time(df,syear,smonth,sday,eyear,emonth,eday):
    #storing a date in a variable in the form of  string
    start=f'{syear}-{smonth}-{sday}'
    #storing a date in a variable in the form of string
    end= f'{eyear}-{emonth}-{eday}'
    #selectin the block between start and end date and storing in mask variable
    mask=(df['Date']>=start)&(df['Date']<=end)
    #getting the mean value between mask timeperiod variable
    daily_ret=df.loc[mask]['daily_return'].mean()
    #storing the selecting period into df2 variable
    df2=df.loc[mask]
    #getting the time period withe the  shape method and saving the days variable
    days=df2.shape[0]
    #getting total returns 
    return (days*daily_ret)
```

```python
#Matplotlib Finance
#creating the function
def mplfinance_plot(ticker,chart_type,syear,smonth,sday,eyear,emonth,eday):
    start=f'{syear}-{smonth}-{sday}'
    #storing a date in a variable in the form of string
    end= f'{eyear}-{emonth}-{eday}'
    try:
        df=pd.read_csv("C:\\Users\\ASUS\\Desktop\\db\\"+ticker+'.csv',index_col='Date',parse_dates=True)
    #writing except block if file don't found
    except FileNotFoundError:
        print("This file did not exist")
    # else part of try except return the file
    else:
        #creating date column as index    
      
    #selecting a particular dataframe
        df_sub=df.loc[start:end]
    #creating candle chart    
        mpf.plot(df_sub,type='candle')
    #creating line chart
        mpf.plot(df_sub,type='line')
    #creating open high low chart
        mpf.plot(df_sub,type='ohlc',mav=4)
    #creating a style and saving to variable s
        s=mpf.make_mpf_style(base_mpf_style='charles',rc={'font.size':8})
    # inserting the style to figure
        fig=mpf.figure(figsize=(12,8),style=s)
    #creating subplot
        ax=fig.add_subplot(2,1,2)
        av=fig.add_subplot(2,1,2,sharex=ax)
        mpf.plot(df_sub,type=chart_type,mav=(3,5,7),ax=ax,volume=av,show_nontrading=True)
    
        
```

```python
mplfinance_plot('AMZN','candle',2020,1,1,2021,1,1)
```

```python
def price_plot(ticker,syear,smonth,sday,eyear,emonth,eday):
    start=f'{syear}-{smonth}-{sday}'
    #storing a date in a variable in the form of string
    end= f'{eyear}-{emonth}-{eday}'
    try:
        df=pd.read_csv("C:\\Users\\ASUS\\Desktop\\db\\"+ticker+'.csv',index_col='Date',parse_dates=True)
    #writing except block if file don't found
    except FileNotFoundError:
        print("This file did not exist")
    # else part of try except return the file
    else:
        #creating date column as index    
        #df.index=-pd.DatetimeIndex(df['Date'])
    #selecting a particular dataframe
        df_sub=df.loc[start:end]
        #saving to numpy array
        df_np=df_sub.to_numpy()
        
        np_adj_close=df_np[:,5]
        date_arr=df_np[:,1]
        fig=plt.figure(figsize=(12,8),dpi=100)
        axes=fig.add_axes([0,0,1,1])
        axes.plot(date_arr,np_adj_close,color='navy')
        axes.xaxis.set_major_locator(plt.MaxNLocator(8))
        axes.grid(True,color='0.6',dashes=((5,2,1,2)))
        axes.set_facecolor('#FAEBD7')
        
```

```python
#calling the function
price_plot('AMZN',2020,1,1,2021,1,1)
```

```python
#creating the function
def download_multiple_stocks(syear,smonth,sday,eyear,emonth,eday,*args):
    for x in args:
        save_to_csv_from_yahoo(x,syear,smonth,sday,eyear,emonth,eday)
        
    
```

```python
def merge_df_by_column_name(col_name,syear,smonth,sday,eyear,emonth,eday,*tickers):
    mult_df=pd.DataFrame()
    start=f'{syear}-{smonth}-{sday}'
    end=f'{eyear}-{emonth}-{eday}'
    for x in tickers:
        mult_df[x]=web.DataReader(x,'yahoo',start,end)[col_name]
    return mult_df
```

```python
#saving a list of stock inside tickers
tickers=['fB','AAPL','NFLX']
#calling the function to download multiple stocks
download_multiple_stocks(2020,1,1,2021,1,1,*tickers)
```

```python
#creating the function
def plot_return_mult_stocks(investment,stock_df):
    (stock_df/stock_df.iloc[0]*investment).plot(figsize=(15,7))
```

```python
#saving list of stock inside a variable
ticker=['FB','AMZN','NFLX','GOOG']
#mering all close place and saving insie a variable
multi_df=merge_df_by_column_name('Adj Close',2020,1,1,2021,1,1,*ticker)
#ploting the return  
plot_return_mult_stocks(100,multi_df)
```

```python
def get_stock_mean_sd(stock_df,ticker):
    return stock_df[ticker].mean(),stock_df[ticker].std()
```

```python
def get_mult_stock_mean_sd(stock_df):
    for stock in stock_df:
        mean,sd=get_stock_mean_sd(stock_df,stock)
        cov=sd/mean
        print('Stock:{:4}Mean:{:7.2f} Standard Deviation:{:2.2f}'.format(stock,mean,sd))
        print('Coefficient of Variation {} n'.format(cov))
```
