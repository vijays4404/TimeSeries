```python
# checking for missing values in the dataset
df.isnull().sum()
```

```python
df.info()
```

```python
#getting column names which has null values
[column for column in df.columns if df[column].isnull().sum()>0]
```

```python
#since it is categorical column filling na with mode
df['Cuisines']=df['Cuisines'].fillna(df['Cuisines'].mode()[0])
```

```python
#again checking for missing values
df.isna().sum()
```

```python
df['Cuisines'].mode()[0]
```

```python
df2
```

```python
#using merge function to add two different dataset
df=pd.merge(df,df2,on='Country Code',how='left')
```

```python
#getting information from one categorical column country with 
df.Country.value_counts()
```

```python
country=df.Country.value_counts().index
```

```python
country_values=df.Country.value_counts().values
```

```python
#creating pie chart in percentage terms
plt.pie(country_values[:3],labels=country[:3],autopct='%1.2f%%')
```

Observation: Zomato maximum records or transaction are from india
    after that USA and United Kingdoms

```python
df.columns
```

```python
#using groupby function to group in different column
rating=df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
rating
```

## Observation

1 When Rating is betwen 4.5 to 4.9 ---> Excellent
2. When Rating are between 4.0 to 3.4-----> Very Good
3. When Rating is between 3.5 to 3.9 -----> Good
4. When Rating is between 3.0 to 3.4 ------> average
5. When Rating is between 2.5 to 2.9 ---->average
6. When Rating is betwen 2.0 to 2.4 -----> Poor

```python
rating.head()
```

```python
#Plotting barplot
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating',y='Rating Count',data=rating)
```

Observation:
    1. Not Rating Count is very high
    2. Maximum number of rating are between 2.5 to 3.4

```python
## Count Plot 
sns.countplot(x='Rating color',data=rating)
```
