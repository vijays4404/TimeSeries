```python
#importing library
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
#loading the dataset
df=pd.read_csv('/kaggle/input/world-university-rankings-202223/2023 QS World University Rankings.csv')
#loading the dataset
df
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
object_columns=[column for column in df.columns if df[column].dtype=='object']
```

```python
object_columns_rank=object_columns[3:]
```

```python
object_columns_rank
```

```python
import numpy as np

for column in object_columns_rank:
    # Step 1: Remove '+' character from values
    df[column] = df[column].apply(lambda x: x.replace('+', '') if isinstance(x, str) and '+' in x else x)

    # Step 2: Convert the column to float if possible
    df[column] = pd.to_numeric(df[column], errors='coerce')  # Set non-numeric values to NaN

    # Step 3: Fill missing values with the mean of the column
    df[column] = df[column].fillna(df[column].mean())
```

```python
df.info()
```

```python
df
```

```python
df.isna().sum()
```

```python
for column in [column for column in df.columns if df[column].dtype!='object'][1:]:
    df[column]=df[column].fillna(df[column].mean())
```

```python
df.isna().sum()
```

```python
for column in [column for column in df.columns if df[column].dtype=='object']:
    encode=LabelEncoder()
    df[column]=encode.fit_transform(df[column])
```

```python
y=df['Rank']
x=df.drop('Rank',axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=123)
```

```python
scaler=StandardScaler()
scaler.fit(x_train)
x_train=pd.DataFrame(scaler.transform(x_train),columns=x_train.columns)
x_test=pd.DataFrame(scaler.transform(x_test),columns=x_test.columns)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
print(model.score(x_test,y_test))
```
