```python
df_1=pd.read_csv('https://raw.githubusercontent.com/derekbanas/TimeSeriesAnalysis/main/owid-covid-data-11-5.csv')
```

```python
df_1
```

```python
#Holt_Winter 
df_2=df_1.location.unique()
indexer=df_1[df_1['location']=='United States'].index
df_2=df_1.loc[indexer,'date':'new_deaths_per_million']
df_2=df_2.dropna()
df_2.set_index('date',inplace=True)
df_2.drop(df_2.columns.difference(['new_deaths_per_million']),1,inplace=True)
df_2.plot(figsize=(12,6))
df_2.rolling(window=30).mean()['new_deaths_per_million'].plot()

```

```python
df_2=df_2.asfreq('d')
df_2.info()
```

```python
train_df=df_2.iloc[:192]
test_df=df_2.iloc[192:]
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
fit_model=ExponentialSmoothing(train_df['new_deaths_per_million'],
                               trend='add',seasonal='add',
                              seasonal_periods=14).fit()
                               
prediction=fit_model.forecast(58)
train_df['new_deaths_per_million'].plot(figsize=(12,6))
test_df['new_deaths_per_million'].plot()
prediction.plot()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
def mae(y1,y2,axis=0):
    y1_np=y1.to_numpy()
    y2_np=y2.to_numpy()
    return np.mean(np.abs(y1_np-y2_np),axis=axis)
```

```python
mae(test_df,prediction,None)
```

```python
mean_absolute_error (test_df,prediction)
```

```python
def mse(y1,y2,axis=0):
    y1_np=y1.to_numpy()
    y2_np=y2.to_numpy()
    return ((y1_np-y2_np)++2).mean(axis=axis)
```

```python
mse(test_df,prediction,None)
```

```python
mean_squared_error(test_df,prediction)
```

```python
#Root Mean Squared Error
np.sqrt(mean_absolute_error (test_df,prediction))
```
