```python
#importing library
import numpy as np
#library for manipulating tabular dataset
import pandas as pd
#train test split to split the dataset into train and test
from sklearn.model_selection import train_test_split
#for scaling the dataset with StandardScaler
from sklearn.preprocessing import StandardScaler
#different type of linear model
from sklearn.linear_model import LinearRegression,Ridge,Lasso
#using KNeighbors
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
#from ensemble class import Random Forest and GradientBoosting
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
import warnings
```

```python
#importing the datatset
df=pd.read_csv('/kaggle/input/ps4-games/games_data.csv')
#showing the dataset
df
```

```python
#dropping the first column
df.drop([df.columns[0],df.columns[1],df.columns[7]],axis=1,inplace=True)
```

```python
#getting the priliminary information
df.info()
```

# Preprocssing

```python
def preprocess_inputs(df):
    df=df.copy()
    y=df['rating']
    x=df.drop('rating',axis=1)
    x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
    #scaling
    scaler=StandardScaler()
    scaler=scaler.fit(x_train)
    x_train=pd.DataFrame(scaler.transform(x_train),columns=x_train.columns)
    x_test=pd.DataFrame(scaler.transform(x_test),columns=x_test.columns)
    
    return x_train,x_test,y_train,y_test
```

```python
x_train,x_test,y_train,y_test=preprocess_inputs(df)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Training the Model

```python
models={
    'Linear Regression': LinearRegression(),
    'Linear Regression(L2 Regularization)': Ridge(),
    'Linear Regression (L1 Regularization)': Lasso(),
    'k-Nearest Neighbors': KNeighborsRegressor(),
    'Neural Network': MLPRegressor(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(),
    'XGBoost': XGBRegressor(),
    'LightGBM': LGBMRegressor(),
    
}
```

```python
#models
models
```

```python
for name,model in models.items():
    model.fit(x_train,y_train)
    print(name+'trained')
```

# Results

```python
for name,model in models.items():
    y_pred=model.predict(x_test)
    rmse=np.sqrt (     np.mean((y_test-y_pred)**2)    )
    print(name+"RMSE:{:.4f}".format(rmse))
```
