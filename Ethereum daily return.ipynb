```python
#importing library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('ethereum.csv')
```

```python
df
```

```python
#visually show and plot the close price
plt.figure(figsize=(14,6))
plt.title('Close Price',fontsize=18)
plt.plot(df['Close'])
plt.xlabel('Date',fontsize=18)
plt.ylabel('Price',fontsize=18)  
plt.show()
```

```python
df['Close'].describe()
```

```python
dsr=df['Close'].pct_change()
```

```python
#show the daily simple return
dsr
```

```python
dsr.describe()
```

```python
#visually plot the daily simple  return
plt.figure(figsize=(12,6))
plt.title('Daily Simple Return')
plt.xlabel('Date',fontsize=18)
plt.ylabel('Percentage',fontsize=18)
plt.plot(dsr.index,dsr)
plt.show()
```

```python
#put the  two charts together
plt.figure(figsize=(12,6))
top=plt.subplot2grid((4,4),(0,0),rowspan=4,colspan=4)
top.plot(df.index,df['Close'],label='Close')
plt.title('Close Price')
plt.legend(loc='upper left')
plt.show()
plt.figure(figsize=(12,6))
bottom=plt.subplot2grid((4,4),(3,0),rowspan=1,colspan=4)
plt.title('DSR')
bottom.plot(dsr.index,dsr)
plt.subplots_adjust(hspace=0.75)
```
