# Creating function to encode Nominal Features

```python
def onehot_encode(df,column):
    #creating the dummy function
    dummies=pd.get_dummies(df[column])
    #concating the column with original dataset
    df=pd.concat([df,dummies],axis=1)
    #dropping the original column
    df=df.drop(column,axis=1)
    return df
    
                              
```

```python
df=onehot_encode(df,'Conference')
```

```python
label_encoder=LabelEncoder()
y=label_encoder.fit_transform(y)
```

```python
scaler=MinMaxScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

# train test split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
x_train.shape
```

```python
inputs=tf.keras.Input(shape=(33,))
x=tf.keras.layers.Dense(16,activation='relu')(inputs)
x=tf.keras.layers.Dense(16,activation='relu')(x)
outputs=tf.keras.layers.Dense(1,activation='sigmoid')(x)
```

```python
model=tf.keras.Model(inputs=inputs,outputs=outputs)
optimizer=tf.keras.optimizers.Adam(learning_rate=0.001)
metrics=[tf.keras.metrics.BinaryAccuracy(name='acc'),
        tf.keras.metrics.AUC(name='auc')]
```

```python
model.compile(optimizer=optimizer,loss='binary_crossentropy',
             metrics=metrics)
batch_size=32
epochs=200
history=model.fit(x_train,y_train,validation_split=0.2,batch_size=batch_size,
                 epochs=epochs,verbose=0)
```

# Results

```python
model.evaluate(x_test,y_test)
```
