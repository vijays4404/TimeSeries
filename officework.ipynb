# Birthday data

```python
dob=pd.to_datetime(data['DOB'])
data['year']=dob.apply(lambda x:x.year)
data['month']=dob.apply(lambda x:x.month)
data['day']=dob.apply(lambda x:x.day)
col = ["month", "day"]
df = data.sort_values(col, ascending=[True, True])
birthday_df=df[['Name','DOB',"Email_Address","Location"]]
birthday_df.to_excel("birthday2.xlsx")
```
