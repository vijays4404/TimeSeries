# Birthday data

```python
dob=pd.to_datetime(data['DOB'])
data['year']=dob.apply(lambda x:x.year)
data['month']=dob.apply(lambda x:x.month)
data['day']=dob.apply(lambda x:x.day)
col = ["month", "day"]
df = data.sort_values(col, ascending=[True, True])
birthday_df=df[['Name','DOB',"Email_Address","Location"]]
birthday_df.to_excel("birthday2.xlsx")
```

```python
date_of_join=data.copy()
doj=date_of_join['DOJ']
date_of_join['year']=doj.apply(lambda x:x.year)
date_of_join['month']=doj.apply(lambda x:x.month)
date_of_join['day']=doj.apply(lambda x:x.day)
joining_df=date_of_join[['Name','DOJ',"Email_Address","Location","year","month"]]
#birthday_df.to_excel("birthday2.xlsx")
date_of_joining=joining_df[joining_df['year']==2022][joining_df['month']==8]

date_of_joining.reset_index(inplace=True)
date_of_joining
```

```python

#create a Pandas Excel writer using XlsxWriter as the engine
writer = pd.ExcelWriter('birthday3.xlsx', engine='xlsxwriter')

#write each DataFrame to a specific sheet
birthday_df.to_excel(writer, sheet_name='birth date')
date_of_joining.to_excel(writer, sheet_name='new joining')


#close the Pandas Excel writer and output the Excel file
writer.save()
```
