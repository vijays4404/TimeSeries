```python
#importing library
#This program use the three moving average crossover strategy to when to buy and sell
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
#loading the data
nifty=pd.read_csv('nifty.csv',index_col='Date',parse_dates=True)
#nifty=nifty.set_index(pd.DatetimeIndex(nifty['Date'].values))
```

```python
#Calculate the three moving averages
#Calculate the short/fast exponential moving average

ShortEMA=nifty.Close.ewm(span=5,adjust=False).mean()
#Calculate the long/slow moving average
LongEMA=nifty.Close.ewm(span=63,adjust=False).mean()
#Calculate the medium moving average
MediumEMA=nifty.Close.ewm(span=21,adjust=False).mean()
```

```python
#vistualize the closing price and the exponential moving average
plt.figure(figsize=(12,5))
plt.title('Close Price',fontsize=18)
plt.plot(nifty['Adj Close'],color='blue')
plt.plot(ShortEMA,label='Short/Fase EMA',color='red')
plt.plot(LongEMA,label='Long/slow EMA',color='green')
plt.plot(MediumEMA,label='Middle/Medium EMA',color='orange')
plt.xlabel('Date',fontsize=18)
plt.ylabel('Close Price',fontsize=18)
plt.show()
```

```python
#creating the function to buy and sell the stock
def buy_sell_function(data):
    buy_list=[]
    sell_list=[]
    flag_long=False
    flag_short=False
    for i in range(0,len(data)):
        if data['Medium'][i]<data['Long'][i] and data['Short'][i]<data['Medium'][i] and flag_long==False and flag_short==False:
            buy_list.append(data['Close'][i])
            sell_list.append(np.nan)
            flag_short=True
        elif flag_short==True and data['Short'][i]>data['Medium'][i]:
            sell_list.append(data['Close'][i])
            buy_list.append(np.nan)
            flag_short=False
        
        elif data['Medium'][i]>data['Long'][i] and data['Short'][i]>data['Medium'][i] and flag_long==False and flag_short==False:
            buy_list.append(data['Close'][i])
            sell_list.append(np.nan)
            flag_short=True
        elif flag_short==True and data['Short'][i]<data['Medium'][i]:
            sell_list.append(data['Close'][i])
            buy_list.append(np.nan)
            flag_short=False
        else:
            buy_list.append(np.nan)
            sell_list.append(np.nan)
    
    return(buy_list,sell_list)
        
```

```python
#Add the buy and sell signals to the data set
nifty['Buy']=buy_sell_function(nifty)[0]
nifty['Sell']=buy_sell_function(nifty)[1]
```

```python
#plotting the graph
plt.figure(figsize=(20,10))
plt.title('Buy and Sell Plot',fontsize=18)
plt.plot(nifty['Adj Close'],color='blue',alpha=0.35)
plt.plot(ShortEMA,label='Short/Fase EMA',color='red',alpha=0.35)
plt.plot(LongEMA,label='Long/slow EMA',color='green',alpha=0.35)
plt.plot(MediumEMA,label='Middle/Medium EMA',color='orange',alpha=0.35)
plt.scatter(nifty.index,nifty['Buy'],color='green',marker='^',alpha=1)

plt.scatter(nifty.index,nifty['Sell'],color='red',marker='v',alpha=1)
plt.xlabel('Date',fontsize=18)
plt.ylabel('Close Price',fontsize=18)
plt.show()
```
