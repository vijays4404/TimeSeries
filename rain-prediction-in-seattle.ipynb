```python
#importing library
import numpy as np
#import pandas
import pandas as pd
#visualization library
import matplotlib.pyplot as plt
import seaborn as sns
#train_test_split
from sklearn.model_selection import train_test_split
#Scaler
from sklearn.preprocessing import StandardScaler
#linear_model
from sklearn.linear_model import LinearRegression,Ridge,LogisticRegression
#KNeighbors
from sklearn.neighbors import KNeighborsRegressor
#tree based model
from sklearn.tree import DecisionTreeRegressor
#Support Vector Machine
from sklearn.svm import LinearSVR,SVR
#Neural Network Regressor
from sklearn.neural_network import MLPRegressor
#ensemble
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
#XGBoost
from xgboost import XGBRegressor
#catboost
from catboost import CatBoostRegressor
#from SVC
from sklearn.svm import LinearSVC,SVC
#accuracy score
from sklearn.metrics import accuracy_score
from lightgbm import LGBMRegressor
```

# Loading the Dataset

```python
df=pd.read_csv('/kaggle/input/did-it-rain-in-seattle-19482017/seattleWeather_1948-2017.csv')
#showing the dataset
df
```

# Checking for Missing Values

```python
def preprocess_inputs(df):
    df=df.copy()
    #dropping the rows with any missing values
    df=df.dropna(axis=0).reset_index(drop=True)
    df['RAIN']=df['RAIN'].astype(np.int)
    #Extracting day, month and year from the date column
    df['Year']=pd.to_datetime(df['DATE']).apply(lambda x:x.year)
    df['Month']=pd.to_datetime(df['DATE']).apply(lambda x:x.month)
    df['Day']=pd.to_datetime(df['DATE']).apply(lambda x:x.day)
    df=df.drop('DATE',axis=1)
    #Splitting into target and feature dataset
    y=df['PRCP']
    x=df.drop("PRCP",axis=1)
    #scaling x
    #train_test_split
    x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,shuffle=True)
    scaler=StandardScaler()
    scaler=scaler.fit(x_train)
    x_train=pd.DataFrame(scaler.transform(x_train),columns=x_train.columns)
    x_test=pd.DataFrame(scaler.transform(x_test),columns=x_test.columns)
    
    return x_train,x_test,y_train,y_test
```

```python
x_train,x_test,y_train,y_test=preprocess_inputs(df)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Creating the dictionary with list of models

```python
models={
    'Linear Regression':LinearRegression(),
    'Ridge':Ridge(),
    'KNeighborsRegressor':KNeighborsRegressor(),
    'Decision Tree':DecisionTreeRegressor(),
    'Linear SVC': LinearSVR(),
    'SVR':SVR(),
    'MLP Regressor':MLPRegressor(),
    'Random Forest':RandomForestRegressor(),
    'Gradient Boosting':GradientBoostingRegressor(),
    'XGB Regressor': XGBRegressor(),
    'LGBM Regressor': LGBMRegressor(),
    'Cat Boost':CatBoostRegressor()}
```

```python
models
```

# Fitting the Model to train set 

```python
for name,model in models.items():
    model.fit(x_train,y_train)
    print(name+':model trained')
```

```python
y_train
```

# Checking Acuracy Score for Each Linear Model

```python
for name,model in models.items():
    print(name)
    print(model.score(x_test,y_test))
```
