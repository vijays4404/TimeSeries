```python
#rename the columns name
df.rename(columns={'ScheduledDay':'ScheduleDate','AppointmentDay':'AppointmentDate'},inplace=True)
```

```python
#extracting year month and day from the date column
df['ScheduledYear']=df['ScheduleDate'].apply(lambda x:np.int(x[0:4]))
df['ScheduledMonth']=df['ScheduleDate'].apply(lambda x:np.int(x[5:7]))
df['ScheduledDay']=df['ScheduleDate'].apply(lambda x:np.int(x[8:10]))
df['ScheduledHour']=df['ScheduleDate'].apply(lambda x:np.int(x[11:13]))
df['ScheduledMinute']=df['ScheduleDate'].apply(lambda x:np.int(x[14:16]))
df['ScheduledSecond']=df['ScheduleDate'].apply(lambda x:np.int(x[17:19]))
```

```python
#dropping the original date column
df
```

```python
df['AppointmentYear']=df['AppointmentDate'].apply(lambda x:np.int(x[0:4]))
df['AppointentMonth']=df['AppointmentDate'].apply(lambda x:np.int(x[5:7]))
df['AppointmentDay']=df['AppointmentDate'].apply(lambda x:np.int(x[8:10]))
```

```python
#dropping the original appointment column
df=df.drop('AppointmentDate',axis=1)
```

```python
#checking for unique value in each object columns
{column:df[column].unique() for column in df.select_dtypes('object').columns}
```

```python
#creating function to encode binary function
def binary_encode(df,column,positive_value):
    df=df.copy()
    #encoding the binary encode 
    df[column]=df[column].apply(lambda x:1 if x==positive_value else 0)
    #returning the dataframe
    return df
```

```python
def onehot_encode(df,column,prefix):
    df=df.copy()
    #creating the dummy
    dummies=pd.get_dummies(df[column],prefix=prefix)
    #concating to original dataframe
    df=pd.concat([df,dummies],axis=1)
    #dropping the original dataframe
    df=df.drop(column,axis=1)
    #returning the dataframe
    return df
    
```

```python
#applying the function to a dataframe
df=binary_encode(df,'Gender',positive_value='M')
df=binary_encode(df,'No-show',positive_value='Yes')
df=binary_encode(df,'Neighbourhood',positive_value='N')
```

```python
df
```

```python
df=df.drop('ScheduleDate',axis=1)
```

```python
df
```

```python
y=df['No-show'].copy()
x=df.drop('No-show',axis=1).copy()
```

```python
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
x
```

```python
#training the dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#Training the model
inputs=tf.keras.Input(shape=(x.shape[1]))
x=tf.keras.layers.Dense(64,activation='relu')(inputs)
x=tf.keras.layers.Dense(64,activation='relu')(x)
outputs=tf.keras.layers.Dense(1,activation='sigmoid')(x)
model=tf.keras.Model(inputs,outputs)
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy',
            tf.keras.metrics.AUC(name='auc')
])
history=model.fit(x_train,y_train,validation_split=0.2,
                    batch_size=32,epochs=100,
                 callbacks=[tf.keras.callbacks.EarlyStopping(
                 monitor='val_loss',
                 patience=3,
                 restore_best_weights=True)])
```

```python
model.evaluate(x_test,y_test)
```

```python
y_true=np.array(y_test)
y_pred=np.squeeze(np.array(model.predict(x_test)>=0.5,dtype=np.int))
```

```python
print('Classification Report',classification_report(y_true,y_pred))
```

```python
print('Confusion_Matrix',confusion_matrix(y_true,y_pred))
```
