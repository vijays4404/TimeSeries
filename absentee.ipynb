```python
#importing library
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import LinearSVR,SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
```

```python
#loading the dataset
df=pd.read_csv('../input/absenteeism-at-work-uci-ml-repositiory/Absenteeism_at_work.csv',delimiter=';')
#showing the dataset
df
```

```python
#Preprocssing the dataset
df.info()
```

# Preprocessing the Dataset

```python
def onehot_encode(df,column,prefix):
    df=df.copy()

    dummies=pd.get_dummies(df[column],prefix=prefix)
    df=pd.concat([df,dummies],axis=1)
    df=df.drop(column,axis=1)
    return df
```

```python
pd.get_dummies(df['Reason for absence'],prefix='Reason')
```

```python
def preprocess_inputs(df):
    df=df.copy()
    #dropping the id column
    df=df.drop('ID',axis=1)
    df=onehot_encode(df,'Reason for absence',prefix='Reason')
    #Split df into x and y
    y=df['Absenteeism time in hours'].copy()
    x=df.drop('Absenteeism time in hours',axis=1).copy()
    #train_test_split
    x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=1)
    scaler= StandardScaler()
    scaler.fit(x_train)
    x_train=pd.DataFrame(scaler.transform(x_train),columns=x_train.columns,index=x_train.index)
    x_test=pd.DataFrame(scaler.transform(x_test),columns=x_test.columns,index=x_test.index)
    
    return x_train,x_test,y_train,y_test
```

```python
x_train,x_test,y_train,y_test=preprocess_inputs(df)
```

# Training the Model

```python
models={
    'Linear Regression': LinearRegression(),
    'K-Nearest Neighbors': KNeighborsRegressor(),
    'Neural Network': MLPRegressor(),
    'Support Vector Machine(Linear)': LinearSVR(),
    'Support Vector Machine(RBF)': SVR(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(),
    'XGBoost': XGBRegressor(),
    'LightGBM':LGBMRegressor(),
    'CatBoost': CatBoostRegressor(verbose=0)
}
```

```python
for name, model in models.items():
    model.fit(x_train,y_train)
    print(name+'trained')
```

# Results

```python
for name,model in models.items():
    print(name+'R^2 Score{:.5f}'.format(model.score(x_test,y_test)))
```
