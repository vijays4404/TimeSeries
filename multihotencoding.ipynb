# Creating the Preprocessing Function

```python
def multihot_encode(df,column):
    df=df.copy()
    df[column]=df[column].apply(lambda x:x.split(','))
    all_categories=np.unique(df[column].sum())
    for category in all_categories:
        df[column+'_'+ category]=df.apply(lambda x:1 if category in df[column] else 0,axis=1)
    df=df.drop(column,axis=1)
    return df
        
    
```

```python
def onehot_encode(df,columns):
    for column in columns:
        dummies=pd.get_dummies(df[column],prefix=column)
        df=pd.concat([df,dummies],axis=1)
        df=df.drop(column,axis=1)
    return df
```

```python
def preprocess_input(df):
    df=df.copy()
    df['Market Category']=df['Market Category'].fillna('Missing')
    #applying the multihot_encode function in the dataset
    df=multihot_encode(df,'Market Category')
    onehot_columns=df.select_dtypes('object')
    df=onehot_encode(df,onehot_columns)
    df['Engine HP']=df['Engine HP'].fillna(df['Engine HP'].mean())
    df['Engine Cylinders']=df['Engine Cylinders'].fillna(df['Engine Cylinders'].mean())
    df['Number of Doors']=df['Number of Doors'].fillna(df['Number of Doors'].mode()[0])
    #splitting betweent target and feature column
    y=df['MSRP']
    x=df.drop('MSRP',axis=1)
    #scaling the dataset
    scaler=StandardScaler()
    x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
    #train_test_split
    x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
    return x_train,x_test,y_train,y_test
```
