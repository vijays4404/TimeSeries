```python
#importing library
import numpy as np
import pandas as pd
import re
import plotly.express as px
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow_addons.metrics import RSquare
```

```python
#importing the data
df=pd.read_csv('../input/finance-accounting-courses-udemy-13k-course/udemy_output_All_Finance__Accounting_p1_p626.csv')
df
```

```python
#checking null values
df.info()
```

```python
#dropping the unneeded column
df.drop(['id','title','url'],axis=1,inplace=True)
```

```python
df
```

```python
#checking for null  values
df.isna().mean()
```

```python
df['discount_price__currency'].unique()
```

```python
df['price_detail__currency'].unique()
```

```python
#dropping unnecessary column
df.drop(['discount_price__currency','price_detail__currency'],axis=1,inplace=True)
```

```python
#checking the  datatypes
df.dtypes
```

```python
#removing non numercial character from the price_detail_colum D is non-numercial character and d is numerical character

(df['discount_price__price_string'].apply(lambda x:np.float(re.sub(r'\D+','',x)) if str(x)!='nan' else x)==df['discount_price__amount']).sum()
```

```python
#dropping column
df.drop(['discount_price__price_string','price_detail__price_string'],axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
#filling na valuues with mean of the column
for column in ['discount_price__amount','price_detail__amount']:
    df[column]=df[column].fillna(df[column].mean())
```

```python
df
```

```python
#creating month and year as new column
df['created_year']=df['created'].apply(lambda x:np.int(x[0:4]))
df['created_month']=df['created'].apply(lambda x: np.int(x[5:7]))

df['published_year']=df['published_time'].apply(lambda x:np.int(x[0:4]))
df['published_month']=df['published_time'].apply(lambda x: np.int(x[5:7]))

                                        
```

```python
#droping created and published_time column
df.drop(['created','published_time'],axis=1,inplace=True)
```

```python
df
```

```python
df.info()
```

```python
#converting to number in is_paid  and is_wishlisted column
df['is_paid']=df['is_paid'].astype(np.int)
df['is_wishlisted']=df['is_wishlisted'].astype(np.int)
```

```python
#selecting target and feature column
y=df['rating'].copy()
x=df.drop('rating',axis=1).copy()
```

```python
#Scaling the feature columns
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

```python
x
```

```python
#spliting the data
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=32)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
#setting up input output layers  the model
inputs=tf.keras.Input(shape=(14,))
x=tf.keras.layers.Dense(64,activation='relu')(inputs)
x=tf.keras.layers.Dense(64,activation='relu')(x)
outputs=tf.keras.layers.Dense(1,activation='linear')(x)
model=tf.keras.Model(inputs,outputs)
```

```python

#compiling the model
model.compile(
    optimizer='adam',
    loss='mse')

batch_size=32
epochs=100
history=model.fit(
    x_train,
    y_train,
    validation_split=0.12,
    batch_size=batch_size,
    epochs=epochs,
    callbacks=[tf.keras.callbacks.ReduceLROnPlateau()])
```

```python
#ploting the data in plotly
fig=px.line(
history.history,
y=['loss','val_loss'],
labels={'x':'epochs','y':'loss'}
)
fig.show()
```

```python
#Result
model.evaluate(x_test,y_test)
```
