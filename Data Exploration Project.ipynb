```python
#importing library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading the dataset
fifa=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/fifa_data.csv')
```

```python
#showing the data
fifa.head()
```

```python
gas=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/gas_prices.csv')
```

```python
gas.head()
```

```python
#ploting the data
plt.title('Gas Prices over Time in USD')

plt.plot(gas.Year,gas.USA,'b.-')

plt.plot(gas.Year,gas.Canada,'r.-')
plt.plot(gas.Year,gas['South Korea'],'g.-')
plt.xticks(gas.Year[::3])
plt.legend()
plt.show()
```

```python
#looping through the country
for country in gas:
    if country !='Year':
        plt.plot(gas.Year,gas[country])
plt.title('Gas Prices over Time in USD')
plt.legend()
plt.show()
```

```python
#getting selected information
countries_to_look_at=['Australia','USA','Canada','South Korea']
for country in  gas:
    if country in countries_to_look_at:
        plt.plot(gas.Year,gas[country])
plt.title('Gas Prices over Time in USD')
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.savefig('Get_Price_figure.png',dpi=300)
plt.show()
```

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
labels=['Left','Right']
plt.pie([left,right],labels=labels,autopct='%.2f%%')

plt.show()
```

```python
fifa.Weight=[int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
fifa.Weight
plt.style.use('ggplot')
light=fifa.loc[fifa.Weight<125].count()[0]
medium=fifa[(fifa.Weight>=150)&(fifa.Weight<175)].count()[0]
heavy=fifa[(fifa.Weight>=200)].count()[0]
labels=['Light','Medium','Heavy']
explode=[.4,.4,.4]
plt.pie([light,medium,heavy],labels=labels,explode=explode,autopct='%.2f%%')
plt.show()
```

```python
#boxplot comparison
barcelona=fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid=fifa.loc[fifa.Club=='Real Madrid']['Overall']
revs=fifa.loc[fifa.Club=='New England Revolution']['Overall']
labels=['FC Barcelona','Real Madrid','New Revolution']
plt.title('Professional Soccer Team Comparison')
plt.ylabel('Fifa Overall Rating')
plt.boxplot([barcelona,madrid,revs])
plt.show()
```
