```python
#importing library
import numpy as np
import pandas as pd
#importing matplotlib library
import matplotlib.pyplot as plt
#LabelEncoder and StandardScaler
from sklearn.preprocessing import StandardScaler,LabelEncoder
#train_test_split
from sklearn.model_selection import train_test_split
#neural network
import tensorflow as tf
```

```python
#loading the dataset
df=pd.read_csv('/kaggle/input/voicegender/voice.csv')
#showing the dataset
df
```

```python
#checking for basic informatiion
df.info()
```

# Using LabelEncoder Class to convert label column

```python
label_encoder=LabelEncoder()
df['label']=label_encoder.fit_transform(df['label'])
```

# Checking Numerical Value for each label

```python
dict(enumerate(label_encoder.classes_))
```

# Splitting and Scaling the Dataset

```python
y=df['label']
x=df.drop("label",axis=1)
```

```python
x
```

# Scaling the Data With StandardScaler

```python
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

# Train Test Split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Model and Training

```python
x.shape
```

# TensorFlow Neural Network

```python
inputs=tf.keras.Input(shape=(20,))
x=tf.keras.layers.Dense(64,activation='relu')(inputs)
x=tf.keras.layers.Dense(64,activation='relu')(x)
outputs=tf.keras.layers.Dense(1,activation='sigmoid')(x)
```

```python
model=tf.keras.Model(inputs=inputs,outputs=outputs)
model.summary()
```

# Compile the Model

```python
model.compile(optimizer='adam',loss='binary_crossentropy',
             metrics=['accuracy',
                     tf.keras.metrics.AUC(name='auc')])
```

# Fitting the Model to the train Dataset

```python
history=model.fit(x_train,y_train,validation_split=0.2,batch_size=32,
                 epochs=100,callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                                    patience=3,restore_best_weights=True)])
```

```python
model.evaluate(x_test,y_test)
```
