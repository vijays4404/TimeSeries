```python
#importing library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading the dataset
df=pd.read_csv('../input/stores-area-and-sales-data/Stores.csv')
#showing the dataset
df
```

```python
#getting information about the dataset
df.info()
```

```python
#checking for the null value in the dataset
df.isna().sum()
```

```python
#dropping the Store ID from the dataset
df=df.drop('Store ID ',axis=1)
```

```python
#importing the regression model
from sklearn.linear_model import LinearRegression
```

```python
#creating instance of the Regression model
model=LinearRegression()
```

```python
#splitting the dataset into x and y
y=df['Store_Sales']
x=df.drop('Store_Sales',axis=1)
```

```python
x
```

```python
#Splting the data best train and test
from sklearn.model_selection import train_test_split
```

```python
#Splitting the data 
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=123)
```

```python
#fitting the model
model.fit(x_train,y_train)
```

```python
#evaluating the model
model.score(x_test,y_test)
```
