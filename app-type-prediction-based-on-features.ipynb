```python
#Importing the library
import numpy as np
import pandas as pd
#visualization 
import matplotlib.pyplot as plt
import seaborn as sns
import re
#scaler
from sklearn.preprocessing import StandardScaler,RobustScaler,MinMaxScaler
from sklearn.model_selection import  train_test_split
#Neural Network
import tensorflow as tf
#Analyzing the Result
from sklearn.metrics import confusion_matrix,classification_report
```

# Loading the Dataset

```python
df=pd.read_csv('/kaggle/input/app-store-apple-data-set-10k-apps/AppleStore.csv')
#showing the dataset
df
```

# Dropping the Unnamed Columns

```python
df=df.drop(df.columns[0],axis=1)
```

```python
descriptions=pd.read_csv('/kaggle/input/app-store-apple-data-set-10k-apps/appleStore_description.csv').loc[:,['id','app_desc']]
descriptions
```

# Merging the df and descriptions Dataset

```python
df=pd.merge(df,descriptions,on='id',how='left')
```

```python
df
```

# Preprocesing the Dataset

```python
def preprocess_inputs(df,scaler):
    df=df.copy()
    #dropping the unecessary columns
    unnecessary_columns=['id','track_name','ver','currency']
    df=df.drop(unnecessary_columns,axis=1)
    #dropping the app_desc
    df=df.drop('app_desc',axis=1)
    #replacing the cont_rating column
    cont_ordering={'4+':4, '12+':12, '17+':17, '9+':9}
    df=df.replace(cont_ordering)
    df['prime_genre']=df['prime_genre'].apply(lambda x:1 if x=='Games' else 0)
    #Splitting between target and feature set
    y=df['prime_genre']
    x=df.drop('prime_genre',axis=1)
    x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
    
    return x,y
```

```python
x,y=preprocess_inputs(df,RobustScaler())
```

```python
y
```

# train test split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y.value_counts()
```

# Training the Neural Network Model

```python
inputs=tf.keras.Input(shape=(11,))
x=tf.keras.layers.Dense(64,activation='relu')(inputs)
x=tf.keras.layers.Dense(64,activation='relu')(x)
outputs=tf.keras.layers.Dense(1,activation='sigmoid')(x)
```

```python
model=tf.keras.Model(inputs=inputs,outputs=outputs)
```

```python
model.summary()
```

# Compiling the Model

```python
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy',tf.keras.metrics.AUC(name='AUC')])
batch_size=32
epochs=50
```

```python
history=model.fit(x_train,y_train,validation_split=0.2,batch_size=batch_size,epochs=epochs,callbacks=[tf.keras.callbacks.ReduceLROnPlateau(),
                                                                                                     tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                                                                                                                           patience=3,
                                                                                                                           restore_best_weights=True)])
```

```python
model.evaluate(x_test,y_test)
```
