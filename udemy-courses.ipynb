```python
#importing numpy 
import numpy as np
#importing pandas 
import pandas as pd
#importing re
import re
#StandardScaler
from sklearn.preprocessing import StandardScaler
#train_test_split
from sklearn.model_selection import train_test_split
#tensorflow
import tensorflow as tf
#RSquare
from  tensorflow_addons.metrics import RSquare
import plotly.express as px
```

```python
#loading the dataset
df=pd.read_csv('/kaggle/input/finance-accounting-courses-udemy-13k-course/udemy_output_All_Finance__Accounting_p1_p626.csv')
#showing the dataset
df
```

```python
#getting information about the dataset
df.info()
```

```python
#dropping unnecessary columns
df=df.drop(['id','title','url'],axis=1)
```

```python
#checking for null values
df.isna().sum()
```

```python
#getting percentage of null value in each column 
df.isna().mean()
```

```python
df=df.drop(['discount_price__price_string','price_detail__price_string'],axis=1)
```

```python
#dropping  the unnecessary columns
df=df.drop(['discount_price__currency','price_detail__currency'],axis=1)
```

```python
df
```

```python
for column in ['discount_price__amount','price_detail__amount']:
    df[column]=df[column].fillna(df[column].mean())
```

```python
#checking for missing values
df.isna().sum()
```

# Feature Engineering

```python
df
```

```python
df['created_year']=df['created'].apply(lambda x:np.int(x[0:4]))
```

```python
df['is_paid'].unique()
```

```python
df['created_month']=df['created'].apply(lambda x:np.int(x[5:7]))


df['published_year']=df['published_time'].apply(lambda x:np.int(x[0:4]))
df['published_month']=df['published_time'].apply(lambda x:np.int(x[5:7]))
df=df.drop(['created','published_time'],axis=1)
```

```python
df
```

```python
df['is_paid']=df['is_paid'].astype(np.int)
```

```python
df['is_wishlisted']=df['is_wishlisted'].astype(np.int)
```

```python
df
```

```python
y=df['rating'].copy()
x=df.drop('rating',axis=1)
```

```python
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

```python
x.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

# Model/Training

```python
inputs=tf.keras.Input(shape=(14,))
x=tf.keras.layers.Dense(64,activation='relu')(inputs)
x=tf.keras.layers.Dense(64,activation='relu')(x)
outputs=tf.keras.layers.Dense(1,activation='linear')(x)
```

```python
outputs
```

```python
model=tf.keras.Model(inputs,outputs)
model.compile(optimizer='adam',loss='mse')
batch_size=32
epochs=100
history=model.fit(x_train,y_train,validation_split=0.12,
                 batch_size=batch_size,epochs=epochs,
                 callbacks=[tf.keras.callbacks.ReduceLROnPlateau()])
```

```python
fig=px.line(history.history,y=['loss','val_loss'],labels={'x':'epochs','y':'loss'})
fig.show()
```

```python
model.evaluate(x_test,y_test)
```
